## Laws of Expectations

Apart from the distributions themselves, a great part of probability theory is concerned with expectations of various functions of the stochastic variable. We present some useful rules and laws of expectations here, as well as some curiosities. Note that although we frequently specify $\mathbb{E}[X]$ and $\mathbb{V}[X]$ as the expectation and variance of $X$, respectively, it should always be kept in mind that these actually denote the expectation and variance of the
distribution $p(X)$ of $X$. Furthermore, $\mathbb{E}[f(X)]$ is the expectation of the quantity $f(X)$ with respect to the distribution of $X$, not the distribution of $f(X)$. We could more properly write this as either $\mathbb{E}_X[f(X)]$ 
or $\mathbb{E}_{p(X)}[f(X)]$ to avoid ambiguity — both forms are prevalent in the literature, along with other variants
such as $\mathbb{E}_p[f(X)]$, et cetera.
However, usually the correct context is clear enough to drop explicit reference to the stochastic variable and/or
its distribution.

### Expectation

Given some continuous stochastic variable $X\in\mathcal{X}$, we define the expectation of $f(X)$ with respect to the distribution $p(X)$ as
\begin{eqnarray}
\mathbb{E}[f(X)] & ~\doteq~ & \int_{\mathcal{X}} f(x)\,p(x)\,dx\,.
\end{eqnarray}
For a discrete variable, the corresponding expectation is defined as
\begin{eqnarray}
\mathbb{E}[f(X)] & ~\doteq~ & \sum_{x\in\mathcal{X}} f(x)\,p(x)\,dx\,,
\end{eqnarray}
instead.
Note that the
[law of the unconcious statistician](https://en.wikipedia.org/wiki/Law_of_the_unconscious_statistician "Wikipedia: Law of the unconcious statistician") (LOTUS) was coined, allegedly by Sheldon Ross, and perhaps humorously, to object to taking this definition axiomatically. Instead, it is claimed, the expectation should be derived theoretically anew for every distribution. However, some statisticians were not amused, and many statisticians apparently accept this definition without question, hence the "law", perhaps.

In practice, we do not expect to encounter any pathological distributions for which the definition fails. Note, however, that we **are** assuming that such expectations exist and are finite — the [Cauchy](https://en.wikipedia.org/wiki/Cauchy_distribution "Wikipedia: Cauchy distribution") distribution is one such exception for which the moments do not exist. 

### Iterated expectations

Given two stochastic variables, $X$ and $Y$, the *unconditional* expectation $\mathbb{E}[X]$ can be defined in terms of the
*conditional* expectation $\mathbb{E}[X\mid Y]$ via
\begin{eqnarray}
\mathbb{E}[X] & ~\doteq~ & \mathbb{E}[\,\mathbb{E}[X\mid Y]\,]\,.
\end{eqnarray}
This is sometimes referred to as the [law of iterated expectations](https://en.wikipedia.org/wiki/Law_of_total_expectation  "Wikipedia: Law of total expectation").
Note that the inner expectation is taken with respect to $X$ (or, actually, the conditional distribution $p(X\mid Y)$) for fixed $Y$,
whereas the outer expectation is taken with respect to variable $Y$ (or $p(Y)$).

As an example, suppose that a sample of values of $X$ may be partitoned into distinct groups based upon the corresponding values of $Y$. Then the law of iterated expectations indicates, informally, that the mean of the sample is the weighted mean of the group means, where the weightings take account of the numbers of samples in the various groups. 

### Variance

The variance, or spread, of the distribution $p(X)$ is defined as
\begin{eqnarray}
\mathbb{V}[X] & ~\doteq~ & \mathbb{E}\left[\left(X-\mathbb{E}[X]\right)^2\right]
\\ & ~=~ &
\mathbb{E}\left[X^2-2X\,\mathbb{E}[X]+\mathbb{E}[X]^2\right]
\\& ~=~ &
\mathbb{E}[X^2]-\mathbb{E}[X]^2\,.
\end{eqnarray}
Alternatively, the rearranged definition, namely
\begin{eqnarray}
\mathbb{E}[X^2] & ~\doteq~ & \mathbb{V}[X] + \mathbb{E}[X]^2\,,
\end{eqnarray}
is often more convenient when computing  moments.

### Iterated variances

Given two stochastic variables, $X$ and $Y$, the *unconditional* variance $\mathbb{V}[X]$ can be defined in terms of the
*conditional* expectation $\mathbb{E}[X\mid Y]$ and variance $\mathbb{V}[X\mid Y]$ via
\begin{eqnarray}
\mathbb{V}[X] & ~\doteq~ & \mathbb{E}[\,\mathbb{V}[X\mid Y]\,] 
+ \mathbb{V}[\,\mathbb{E}[X\mid Y]\,]
\,.
\end{eqnarray}
This is sometimes referred to as the [law of iterated variances](https://en.wikipedia.org/wiki/Law_of_total_variance  "Wikipedia: Law of total variance").
Gelman *et al.* [[GCSR1995]](references.ipynb#GCSR1995 "References: Bayesian Data Analysis") offer a short proof, given here in reverse:
\begin{eqnarray}
\mathbb{V}[X] & ~=~ & \mathbb{E}[X^2]-\mathbb{E}[X]^2
\nonumber\\& ~=~ &
\mathbb{E}[\,\mathbb{E}[X^2\mid Y]\,]
-\mathbb{E}[\,\mathbb{E}[X\mid Y]\,]^2
\nonumber\\& ~=~ &
\mathbb{E}\left[\,\mathbb{E}[X^2\mid Y]-\mathbb{E}[X\mid Y]^2\,\right]
+
\mathbb{E}[\,\mathbb{E}[X\mid Y]^2\,]-\mathbb{E}[\,\mathbb{E}[X\mid Y]\,]^2
\nonumber\\& ~=~ &
\mathbb{E}[\,\mathbb{V}[X\mid Y]\,] + \mathbb{V}[\,\mathbb{E}[X\mid Y]\,]
\,.
\end{eqnarray}

To continue with our [previous](#Iterated-expectations "Section: Iterated expectations") example, 
the first term on the right-hand side is the (weighted) mean of the group variances, and represents the
overall *within-group* variance. The second term on the right-hand side is the (weighted) variance of the group means, and represents the *between-group* variance.
Hence, the law of iterated variances informally states that the total variance is the sum of the within-group variance and the between-group variance. This is useful, for example, when measuring the compactness of clustering data points into groups. The within-group variance measures the dissimilarity between points within the same cluster, whereas the between-group variance measures the dissimilarity between the centroids of different clusters. A good clustering solution ahould attempt to minimise the former and maximise the latter.

### Covariance

Given two stochastic variables, $X$ and $Y$, the covariance between the distributions $p(X)$ and $p(Y)$
is given by
\begin{eqnarray}
\texttt{Cov}[X,Y] & ~\doteq~ & \mathbb{E}\left[\left(X-\mathbb{E}[X]\right)\,\left(Y-\mathbb{E}[Y]\right)\right]
\\ & ~=~ &
\mathbb{E}\left[\,XY-X\,\mathbb{E}[Y]-Y\,\mathbb{E}[X]+\mathbb{E}[X]\,\mathbb{E}[Y]\,\right]
\\& ~=~ & \mathbb{E}[XY]-\mathbb{E}[X]\,\mathbb{E}[Y]\,.
\end{eqnarray}
Noted that it [follows](#Variance "Section: Variance") that $\mathbb{V}[X]=\mathtt{Cov}[X,X]$.

### Law of iterated covariances

Given three stochastic variables, $X$, $Y$ and $Z$, the *unconditional* covariance $\mathtt{Cov}[X,Y]$ can be defined in terms of *conditional* expectations and covariance via
\begin{eqnarray}
\mathtt{Cov}[X,Y] & ~\doteq~ & \mathbb{E}[\,\mathtt{Cov}[X,Y\mid Z]\,] + 
\mathtt{Cov}[\,\mathbb{E}[X\mid Z],\mathbb{E}[Y\mid Z]\,]
\,.
\end{eqnarray}
This is sometimes referred to as the [law of iterated covariances](https://en.wikipedia.org/wiki/Law_of_total_covariance  "Wikipedia: Law of total covariance").


### Expectations of extremes

Given two stochastic variables, $X$ and $Y$, the *order statistics* that summarise the extreme values are given by
$X_{(1)}\doteq\min(X,Y)$ and $X_{(2)}\doteq\max(X,Y)$, such that $X_{(1)}\le X_{(2)}$. It can then be shown that
\begin{eqnarray}
 & X+Y & ~=~ \min(X,Y)+\max(X,Y)
\\
\Rightarrow~ & \mathbb{E}[X]+\mathbb{E}[Y] &~=~ \mathbb{E}[\min(X,Y)]+\mathbb{E}[\max(X,Y)]
\\
\Rightarrow~ & \mathbb{E}[\max(X,Y)] & ~=~ \mathbb{E}[X]+\mathbb{E}[Y] - \mathbb{E}[\min(X,Y)]\,.
\end{eqnarray}
This is sometimes written in the form
\begin{eqnarray}
\mathbb{E}[X\lor Y] & ~=~ \mathbb{E}[X]+\mathbb{E}[Y] - \mathbb{E}[X\land Y]\,,
\end{eqnarray}
where the operators '$\land$' and '$\lor$' are taken to represent minimisation and maximisation, respectively.
This form is chosen to correspond to the usual law of probabilities, i.e. either that
\begin{eqnarray}
p(X\cup Y) & ~=~ p(X)+p(Y) - p(X\cap Y)\,,
\end{eqnarray}
for sets $X$ and $Y$, where '$\cup$' and '$\cap$' represent set union and intersection, respectively, or else
\begin{eqnarray}
p(X\lor Y) & ~=~ p(X)+p(Y) - p(X\land Y)\,,
\end{eqnarray}
where $X$ and $Y$ are Boolean predicates, and '$\lor$' and '$\land$' represent logical disjunction and conjunction, 
respectively.
However, due to the ambiguity between operator definitions, I personally prefer the version in terms of explicit maximisation and minimisation.

### Mean field approximation

Given a stochastic variable $X\in\mathcal{X}$, consider a transformation 
$f:\mathcal{X}\mapsto\mathcal{Y}$, such that
$Y=f(X)\in\mathcal{Y}$.
Assuming that the function $f$ is sufficiently differentiable, then the Taylor series expansion of $f$ around
the mean $\mathbb{E}[X]\in\mathcal{X}$ is given by
\begin{eqnarray}
& f(X) & ~=~ f\left(\mathbb{E}[X]\right) 
+ f'\left(\mathbb{E}[X]\right)\,\left(X-\mathbb{E}[X]\right)
+\frac{1}{2}f''\left(\mathbb{E}[X]\right)\,\left(X-\mathbb{E}[X]\right)^2+\cdots
\\
\Rightarrow~ & \mathbb{E}[f(X)] & ~=~ f\left(\mathbb{E}[X]\right)
+ \frac{1}{2}f''\left(\mathbb{E}[X]\right)\,\mathbb{V}[X]+\cdots\,.
\end{eqnarray}
Provided that the terms on the right-hand side after the first term are negligable, then we
obtain the so-called *mean field approximation* (MFA):
\begin{eqnarray}
\mathbb{E}[f(X)] & ~\approx~ f\left(\mathbb{E}[X]\right)\,.
\end{eqnarray}
Note that we are using the word 'field' here in the sense of physics and not mathematics, e.g. an electric field
with a strength that varies in the spatial coordinates and time.
Informally, the MFA states that the mean of a function is approximately a function of the mean.

We can also approximate the variance of the tansformation via
\begin{eqnarray}
\mathbb{V}[f(X)] & ~=~ & \mathbb{E}\left[\left(f(X)-\mathbb{E}[f(X)]\right)^2\right]
\\& ~\approx~ & \mathbb{E}\left[f'(\mathbb{E}[X])^2\,\left(X-\mathbb{E}[X]\right)^2\right]
\\ & ~=~ & f'(\mathbb{E}[X])^2\,\mathbb{V}[X]\,,
\end{eqnarray}
provided that $f'(\mathbb{E}[X])\ne 0$. If this condition does not hold, then we can include an extra term
from the Taylor series expansion,
such that
\begin{eqnarray}
\mathbb{V}[f(X)] & ~\approx~ & \frac{1}{4}f''(\mathbb{E}[X])^2\,\mathbb{V}\left[\left(X-\mathbb{E}[X]\right)^2\right]\,,
\end{eqnarray}
instead, although this will require more work to evaluate.
