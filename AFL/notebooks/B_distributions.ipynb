## Beta-Bernoulli Distribution

Consider the situation where the outcome $X$ (i.e. a win or loss) of a match between team A and team B is stochastically decided,
given some (unknown) probability $\theta$ that team A wins. Essentially, the outcome may be considered as a biased coin toss.
However, prior to the match we might consider the relative strengths of teams A and B, and this knowledge might inform us
as to likely values of the probability $\theta$. Consequently, we may model $X$ by a continuous mixture distribution over 
$\theta$, and then use this to derive a posterior distribution for $\theta$.

### Bernoulli data likelihood

Let the variate $X\in\{0,1\}$ represent the outcome of a single match between some arbitrary team A and team B, with $X=1$ indicating a win by team A, and $X=0$ indicating a win by team B.
For convenience, we consider the Bernoulli distribution
\begin{eqnarray}
X\mid\theta & ~\sim~ & \texttt{Bern}(\theta)\,,
\end{eqnarray}
where $\theta$ is the assumed probability that team A wins, which is decided before the match commences.
Thus, we see that
\begin{eqnarray}
p(X=x\mid\theta) & ~=~ & \theta^x\,(1-\theta)^{1-x}\,.
\end{eqnarray}
If we further wish to model the outcomes of an ordered sequence of $n$ *conditionally independent* matches between the same teams A and B, 
with the same parameter
$\theta$, then we will obtain the joint data likelihood
\begin{eqnarray}
p(X=w\mid n,\theta) & ~=~ & \theta^w\,(1-\theta)^{n-w}\,,
\end{eqnarray}
where $w\in\{0,1,\ldots,n\}$ represents the total number of wins for team A, and $\ell=n-w$ represents the total number of
losses.

Note that the use of the Bernoulli distribution explicitly presumes that only wins and losses are possible.
How then might we deal with games for which matches may end in a draw?
It turns out that there are a number of good reasons to treat a draw as being half-a-win and half-a-loss for each team.
Thus, we might define the weighted log-likelihood for a draw as being given by
\begin{eqnarray}
L_\texttt{draw}(\theta) & ~\doteq~ & \frac{1}{2}\ln p(X=1\mid\theta)+\frac{1}{2}\ln p(X=0\mid\theta)
\\& =  &
\frac{1}{2}\ln\theta+\frac{1}{2}\ln\,(1-\theta)
~=~
\ln\left[\theta^{\frac{1}{2}}\,(1-\theta)^{\frac{1}{2}}\right]\,.
\end{eqnarray}
We observe that functionally this takes the same form as
\begin{eqnarray}
L_\texttt{draw}(\theta) & ~=~ & \ln p(X=0.5\mid\theta)\,.
\end{eqnarray}
Consequently, we may numerically treat observed draws as having the value $X=\frac{1}{2}$.

### Beta prior distribution

We seek a non-informative prior distribution for $\theta$. 
Following Box and Tiao [[1]](#Citations "Citation [1]: Bayesian Inference in Statistical Analysis"), 
we desire a transformation $\phi(\theta)$ such that the likelihood $p(X=w\mid n,\theta)$, plotted as a function of $\phi$, remains
approximately invariant in shape and size for some fixed $n$ as $w$ varies (but varies in its mean location). 
A uniform prior for $\phi$ then induces a
non-informative prior for $\theta$. In general, it turns out that the relevant prior is usually inversely
proportional to the square-root of the variance. For the Bernoulli distribution, we have
\begin{eqnarray}
\mathbb{V}[X\mid\theta] & ~=~ \theta\,(1-\theta)\,,
\end{eqnarray}
whereupon the non-informative prior is therefore
\begin{eqnarray}
p(\theta) ~\propto~ \frac{1}{\sqrt{\theta\,(1-\theta)}}
& ~~~\Rightarrow~~~ & \theta~\sim~\mathtt{Beta}\left(\frac{1}{2},\frac{1}{2}\right)
\,.
\end{eqnarray}
Observe that this is just 
[Jeffreys' prior](https://en.wikipedia.org/wiki/Jeffreys_prior "Wikipeida: Jeffreys' prior"), 
which notionally corresponds to a single prior pseudo-match
with half-a-win and half-a-loss, i.e. a draw.

More generally, we might choose an arbitrary beta prior distribution
\begin{eqnarray}
\theta & ~\sim~ & \mathtt{Beta}(\alpha,\beta)
\,,
\end{eqnarray}
for some hyper-parameters $\alpha$ and $\beta$, where
$\alpha$ represents the prior number of pseudo-wins and $\beta$ represents the prior number of pseudo-losses.
Note that technically the parameter dependence is explicitly $\theta\mid\alpha,\beta$. However, by convention
the hyper-parameters are often left implicit to indicate that they are held constant. Despite this useful convention,
it is best to always explicitly condition on all relevant parameters when dealing with the actual probability functions,
in order to avoid accidental ambiguity, e.g. by hiding important modelling assumptions.

The special case of $\alpha=\beta$ corresponds to having no other prior information with which to
preference a win/loss over a loss/win.
We could, for example, choose the 
[Haldane prior](https://en.wikipedia.org/wiki/Beta_distribution "Wikipedia: Beta distribution") 
with $\alpha=\beta=0$ to indicate that we have no
prior pseudo-matches.
Similarly, the uniform prior with $\alpha=\beta=1$ corresponds to one previous pseudo-win and one pseudo-loss.
Alternatively, 
Kerman [[2]](#Citations "Citation [2]: Neutral noninformative and informative conjugate beta and gamma prior distributions") 
derives $\alpha=\beta=\frac{1}{3}$ as being a "*neutral*" prior that has the property that it leads "to posterior distributions with approximately 50 per cent probability that the true value is either smaller or larger than the maximum likelihood estimate".

Note that the beta distribution is also a good choice due to its complementary symmetry in $\alpha$ and $\beta$.
Suppose that $\alpha$ denotes our pre-match knowledge about the strength of team A, and similarly let
$\beta$ represent our prior knowlegde about team B. Now, $\theta=\theta_A$ is the prior probability of team A winning,
such that $\theta_A\sim\texttt{Beta}(\alpha,\beta)$.
However, the ordering of teams A and B is arbitrary. If we swapped the order of the teams and their respective hyper-parameters, 
then we would instead have $\theta_B\sim\texttt{Beta}(\beta,\alpha)$.
The beta distribution correctly satisfies the constraint that
\begin{eqnarray}
\mathbb{E}[\theta\mid\alpha,\beta] & ~=~ \frac{\alpha}{\alpha+\beta}~=~1-\mathbb{E}[\theta\mid\beta,\alpha]\,. 
\end{eqnarray}
This is the aforementioned complementary symmetry.

### Beta-Bernoulli mixture distribution

The mixture distribution for a single match is now derived as
\begin{eqnarray}
p(X=x\mid\alpha,\beta) & ~=~ & \int_0^1 p(x\mid\theta)\,p(\theta\mid\alpha,\beta)\,d\theta
\\& = &
\int_0^1 \theta^x(1-\theta)^{1-x}\,
\frac{\theta^{\alpha-1}(1-\theta)^{\beta-1}}{B(\alpha,\beta)}\,d\theta
\\& = &
\frac{1}{B(\alpha,\beta)}\int_0^1 \theta^{\alpha+x-1}(1-\theta)^{\beta-x}\,d\theta
\\& =  &
\frac{B(\alpha+x,\beta-x+1)}{B(\alpha,\beta)}\int_0^1 
\frac{\theta^{\alpha+x-1}(1-\theta)^{\beta-x}}{B(\alpha+x,\beta-x+1)}\,d\theta\,.
\end{eqnarray}
Recognising the final integrand as the $\texttt{Beta}(\alpha+x,\beta-x+1)$ distribution, we obtain
\begin{eqnarray}
p(X=x\mid\alpha,\beta) & ~=~ &
\frac{B(\alpha+x,\beta-x+1)}{B(\alpha,\beta)}\,.
\end{eqnarray}
Next, we expand $B(\cdot,\cdot)$ in terms of the *gamma* function $\Gamma(\cdot)$ to obtain
\begin{eqnarray}
p(X=x\mid\alpha,\beta) & ~=~ &
\frac{\Gamma(\alpha+x)\,\Gamma(\beta+1-x)}{\Gamma(\alpha+\beta+1)}\,
\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\,\Gamma(\beta)}\,.
\end{eqnarray}
Finally, using the recurrence relation that $\Gamma(z+1)=z\,\Gamma(z)$, we deduce that
\begin{eqnarray}
p(X=1\mid\alpha,\beta)~=~\frac{\alpha}{\alpha+\beta}\,, & \;\;\;\; &
p(X=0\mid\alpha,\beta)~=~\frac{\beta}{\alpha+\beta}\,,
\end{eqnarray}
such that
\begin{eqnarray}
p(X=x\mid\alpha,\beta) & ~=~ & \frac{\alpha^x\,\beta^{1-x}}{\alpha+\beta}\,.
\end{eqnarray}
Observe that this is just
\begin{eqnarray}
X\mid\bar{\theta} & ~\sim~ & \texttt{Bern}(\bar{\theta})\,,
\end{eqnarray}
with $\bar{\theta}=\frac{\alpha}{\alpha+\beta}$ being the mean of the beta distribution.

Subsequently, for the case of $n$ *unconditionally independent* matches between teams A and B, we would obtain the joint distribution
\begin{eqnarray}
p_u(X=w\mid n,\alpha,\beta) & ~=~ & \frac{\alpha^w\,\beta^{n-w}}{(\alpha+\beta)^n}~=~\bar{\theta}^w\,(1-\bar{\theta})^{n-w}\,.
\end{eqnarray}
Note that this is notionally different from the *conditionally independent* case assumed
[previously](#Bernoulli-data-likelihood "Section: Bernoulli data likelihood"),
where the $n$ matches all shared the same value of the parameter $\theta$. In the unconditional form, the parameter
$\theta$ is assumed to be resampled before every match.

Consequently, for the conditionally independent case, we would instead derive that
\begin{eqnarray}
p_c(X=w\mid n,\alpha,\beta) & ~=~ & \int_0^1 p(X=w\mid n,\theta)\,p(\theta\mid\alpha,\beta)\,d\theta
\\& = &
\int_0^1 \theta^w\,(1-\theta)^{n-w}\,
\frac{\theta^{\alpha-1}(1-\theta)^{\beta-1}}{B(\alpha,\beta)}\,d\theta
\\& = &
\frac{1}{B(\alpha,\beta)}\int_0^1 \theta^{\alpha+w-1}(1-\theta)^{\beta+n-w-1}\,d\theta
\\& =  &
\frac{B(\alpha+w,\beta+n-w)}{B(\alpha,\beta)}\,.
\end{eqnarray}

### Beta posterior distribution

For the case of observing a sequence of $n$ conditionally independent matches, the posterior distribution is given by
\begin{eqnarray}
p(\theta\mid n,w,\alpha,\beta) & ~=~ &
\frac{p(X=w\mid n,\theta)\,p(\theta\mid\alpha,\beta)}{p_c(X=w\mid n,\alpha,\beta)}
~=~
\frac{\theta^{\alpha+w-1}(1-\theta)^{\beta+n-w-1}}{B(\alpha+w,\beta+n-w)}\,,
\end{eqnarray}
such that
\begin{eqnarray}
\theta\mid n,w & ~\sim~ & \texttt{Beta}(\alpha+w,\beta+n-w)\,.
\end{eqnarray}

### Beta-Bernoulli predictive distribution

Suppose the result of the $(n+1)$-th match is now $X$, conditionally independent of previous matches for the same
parameter $\theta$. Then the likelihood of this result is
\begin{eqnarray}
p(X=x\mid\theta) & ~=~ & \theta^{x}\,(1-\theta)^{1-x}\,.
\end{eqnarray}
Hence, the predictive probability of this result is given by
\begin{eqnarray}
p(X=x\mid n,w,\alpha,\beta) & ~=~ & \int_0^1 p(x\mid\theta)\,p(\theta\mid n,w,\alpha,\beta)\,d\theta
\\&~=~&
\int_0^1\frac{\theta^{\alpha+w+x-1}\,(1-\theta)^{\beta+n-w-x}}
{B(\alpha+w,\beta+n-w)}\,d\theta
\\ & = &
\frac{B(\alpha+w+x,\beta+n-w+1-x)}
{B(\alpha+w,\beta+n-w)}
\,.
\end{eqnarray}

In terms of the gamma function, $\Gamma(\cdot)$, this becomes
\begin{eqnarray}
p(X=x\mid n,w,\alpha,\beta) & ~=~ & 
\frac{\Gamma(\alpha+w+x)\,\Gamma(\beta+n-w+1-x)}
     {\Gamma(\alpha+\beta+n+1)}\,
\frac{\Gamma(\alpha+\beta+n)}
     {\Gamma(\alpha+w)\,\Gamma(\beta+n-w)}\,.
\end{eqnarray}
For a loss, i.e. $X=0$, the respective probability reduces to
\begin{eqnarray}
p(X=0\mid n,w,\alpha,\beta) & ~=~ & \frac{\beta+n-w}{\alpha+\beta+n}\,,
\end{eqnarray}
again using the recurrence relation that $\Gamma(z+1)=z\,\Gamma(z)$.
The corresponding probability of a win is therefore
\begin{eqnarray}
p(X=1\mid n,w,\alpha,\beta) & ~=~ & \frac{\alpha+w}{\alpha+\beta+n}\,,
\end{eqnarray}
such that
\begin{eqnarray}
p(X=x\mid n,w,\alpha,\beta) & ~=~ & \frac{(\alpha+w)^x\,(\beta+n-w)^{1-x}}{\alpha+\beta+n}\,.
\end{eqnarray}
Defining $\hat{\theta}\doteq\frac{\alpha+w}{\alpha+\beta+n}$, we observe that
\begin{eqnarray}
p(X=x\mid n,w,\alpha,\beta) & ~=~ & \left(\hat{\theta}\right)^x\,\left(1-\hat{\theta}\right)^{1-x}
\,,
\end{eqnarray}
such that
\begin{eqnarray}
X\mid\hat{\theta} & ~\sim~ & \texttt{Bern}(\hat{\theta})
\,.
\end{eqnarray}

Note that the denominator of $\hat{\theta}$ corresponds to assuming $n$ observed matches plus $\alpha+\beta$ prior pseudo-matches, and the numerator corresponds to $w$ observed wins plus $\alpha$ prior pseudo-wins.
Thus, $\hat{\theta}$ is just a *smoothed* estimate of the empirical probability of a win.

## Beta-Binomial Distribution

The binomial distribution may be derived as a sum of $n$ independent
Bernoulli variables. Hence, we may generalise the
[beta-Bernoulli](#Beta-Bernoulli-Distribution "Section: Beta-Bernoulli Distribution") distribution to the binomial case.
One rationale for doing so is to model  data that are empirically
over-dispersed with respect to the binomial distribution.
This is similar to our later use of the
[gamma-Poisson](#Gamma-Poisson-Distribution "Section: Gamma-Poisson Distribution") distribution to handle over-dispersion with respect to the
Poisson distribution.

### Beta prior distribution (again)

Once again, we utilise the general
[beta](#Beta-prior-distribution "Section: Beta prior distribution")
prior of
\begin{eqnarray}
\theta & ~\sim~ & \mathtt{Beta}(\alpha,\beta)
\,,
\end{eqnarray}
for which the mean and variance are given by
\begin{eqnarray}
\mu_\theta & ~\doteq~ & \mathbb{E}[\theta\mid\alpha,\beta]
~=~\frac{\alpha}{\alpha+\beta}\,,
\\
\sigma^2_\theta & ~\doteq~ & \mathbb{V}[\theta\mid\alpha,\beta]
~=~\frac{\alpha\beta}{(\alpha+\beta)^2\,(\alpha+\beta+1)}\,,
\end{eqnarray}
respectively.
[Recall](https://en.wikipedia.org/wiki/Beta_distribution
"Wikipedia: Beta distribution") that the beta density function is
\begin{eqnarray}
p(\theta\mid\alpha,\beta) & ~=~ &
\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\,\Gamma(\beta)}\,
\theta^{\alpha-1}\,(1-\theta)^{\beta-1}\,.
\end{eqnarray}


### Binomial data likelihood

The binomial distribution abstracts from each observed Bernoulli sequence of $n$ trials by disregarding the order and simply counting the total number of
special events of interest (e.g. successes, or failures) which occurred. It happens that such sequences are *exchangeable*, meaning that every perturbation of an observed
sequence has the *same* probability of occurrence. Thus, there are
$n\choose x$ exchangeable sequences all having exactly $x$ special events and $n-x$ non-special events, in any order.

Thus, let the count $X$ follow a binomial distribution, namely
\begin{eqnarray}
X\mid\theta & ~\sim~ & \texttt{Binom}(n,\theta)\,,
\end{eqnarray}
where $\theta$ is the probability of an event of interest; by convention, we may make the dependence upon $n$ implicit since it is assumed to be a constant
hyper-parameter.
In our sporting match scenario, we might let $n$ be the total allowable
number of attempted scoring shots in a fixed period of time.
Then $\theta$ might be the probability 
that any given attempt actually succeeds. Thus, $X$ would give the total score
of a given team for the match.

The likelihood of a given score is then
\begin{eqnarray}
p(X=x\mid n,\theta) & ~=~ & {n\choose x}\,\theta^x\,(1-\theta)^{1-x}\,,
\end{eqnarray}
which gives a conditional mean and variance of
\begin{eqnarray}
\mu_{X\mid\theta} & ~\doteq~ & \mathbb{E}[X\mid n,\theta]
~=~n\,\theta\,,
\\
\sigma^2_{X\mid\theta} & ~\doteq~ & \mathbb{V}[X\mid n,\theta]
~=~ n\,\theta\,(1-\theta)\,,
\end{eqnarray}
respectively. Observe that $\sigma^2_{X\mid\theta}<\mu_{X\mid\theta}$ for all $\theta\in(0,1)$.
Thus the binomial distribution is under-dispersed in comparison to
the Poisson distribution, for which 
$\sigma^2_{X\mid\lambda}=\mu_{X\mid\lambda}=\lambda$.

### Beta-binomial mixture distribution

To allow for over-dispersion, we permit the probability $\theta$
to first be drawn from the beta prior before the start of each match, and then to be held constant for the duration of the match.
The law of 
[iterated expectations](https://en.wikipedia.org/wiki/Law_of_total_expectation
"Wikipedia: Law of total expectation") then gives
\begin{eqnarray}
\mu_X & ~\doteq~ & \mathbb{E}[X\mid\alpha,\beta] ~=~
\mathbb{E}[\mu_{X\mid\theta}\mid\alpha,\beta]
\\
& ~=~ & \mathbb{E}[n\,\theta\mid\alpha,\beta]~=~n\mu_\theta\,.
\end{eqnarray}
Similarly, the 
law of 
[iterated variances](https://en.wikipedia.org/wiki/Law_of_total_variance
"Wikipedia: Law of total variance")
gives
\begin{eqnarray}
\sigma^2_X & ~\doteq~ & \mathbb{V}[X\mid\alpha,\beta] ~=~
\mathbb{V}[\mu_{X\mid n,\theta}\mid\alpha,\beta]
+
\mathbb{E}[\sigma^2_{X\mid n,\theta}\mid\alpha,\beta]
\\
& ~=~ & 
\mathbb{V}[n\,\theta\mid\alpha,\beta]
+
\mathbb{E}[n\,\theta\,(1-\theta)\mid\alpha,\beta]
\\
& ~=~ & n^2\sigma^2_\theta+n\,\mu_\theta-n\,(\sigma^2_\theta+\mu_\theta^2)
\\
& = &
n\,\mu_\theta\,(1-\mu_\theta)+n(n-1)\,\sigma^2_\theta
\,.
\end{eqnarray}
Observe that if $\sigma^2_\theta=0$ then $\theta=\mu_\theta$ and the variance reduces to the usual binomial variance of $\sigma^2_X=n\,\theta\,(1-\theta)$.
Hence, for $\sigma^2_\theta>0$ the beta-binomial allows for additional
variance over the binomial distribution.

The 
[beta-binomial](https://en.wikipedia.org/wiki/Beta-binomial_distribution
"Wikipedia: Beta-binomial distribution")
mixture distribution for a single score $X$ is now derived as
\begin{eqnarray}
p(X=x\mid n,\alpha,\beta) & ~=~ & \int_0^1 p(x\mid n,\theta)\,p(\theta\mid\alpha,\beta)\,d\theta
\\& = &
\int_0^1 {n\choose x}\theta^x(1-\theta)^{n-x}\,
\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\,\Gamma(\beta)}
\,\theta^{\alpha-1}(1-\theta)^{\beta-1}\,d\theta
\\& = &
{n\choose x}\,\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\,\Gamma(\beta)}
\int_0^1 \theta^{\alpha+x-1}(1-\theta)^{\beta+n-x-1}\,d\theta
\\& =  &
{n\choose x}\,\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\,\Gamma(\beta)}
\,\frac{\Gamma(\alpha+x)\,\Gamma(\beta+n-x)}{\Gamma(\alpha+\beta+n)}
\,,
\end{eqnarray}
since the final integrand is just the $\texttt{Beta}(\alpha+x,\beta+n-x)$ kernel.
The mean and variance of this distribution are given by
\begin{eqnarray}
\mu_X~=~\frac{n\alpha}{\alpha+\beta}\,, & ~~~~\mbox{ }~~~ &
\sigma_X^2~=~\frac{n\alpha\beta\,(\alpha+\beta+n)}
{(\alpha+\beta)^2\,(\alpha+\beta+1)}\,,
\end{eqnarray}
respectively. Consequently, defining $\nu\doteq\alpha+\beta$, we
obtain
\begin{eqnarray}
\sigma^2_X & ~=~ & 
\frac{\nu+n}{\nu+1}\,\left(\mu_X-\frac{1}{n}\mu_X^2\right)
\end{eqnarray}
as the mean-to-variance relation.

Let us now define $z!\doteq\Gamma(z+1)$ for real-valued $z\in[0,\infty)$.
Then the beta-binomial distribution becomes
\begin{eqnarray}
p(X=x\mid n,\alpha,\beta) & ~=~ &
\frac{n!\,(\alpha+\beta-1)!\,(\alpha+x-1)!\,(\beta+n-x-1)!}
{x!\,(n-x)!\,(\alpha-1)!\,(\beta-1)!\,(\alpha+\beta+n-1)!}
\\
& = &
\frac{(\alpha+x-1)!}{x!\,(\alpha-1)!}\,
\frac{(\beta+n-x-1)!}{(n-x)!\,(\beta-1)!}
\left/\frac{(\alpha+\beta+n-1)!}{n!\,(\alpha+\beta-1)!}\right.
\\ & = &
{\alpha+x-1\choose x}\,{\beta+n-x-1\choose n-x}
\left/{\alpha+\beta+n-1\choose n}\right.\,.
\end{eqnarray}
This is just the real-valued  version of
the [negative hypergeometric](https://en.wikipedia.org/wiki/Negative_hypergeometric_distribution
"Wikipedia: Negative hypergeometric distribution")
distribution, i.e. a *Polya urn* distribution.
Notionally, in the integer-valued version, the urn contains $K=n$ successes and $N-K=\alpha+\beta-1$ failures,
and sampling without replacement halts as soon as $r=\alpha$ failures have been observed, giving $k=x$ observed successes.

Note that, paradoxically, $\alpha$ is actually the prior number of *successes*
from the $\texttt{Beta}(\alpha,\beta)$ prior distribution.
The fact that $r=\alpha$ is also the required number of urn sampling *failures* can be explained
(with a bit of hand waving)
by the fact that a high value of $\alpha$ would imply that a high number $x$ of successes would be expected to occur
before $\alpha$ failures were finally observed.
This makes sence since, from the beta distribution, a high value of $\alpha$ would lead to a high expected probability of observing a success in any given
Bernoulli trial.

### Joint binomial processes

Now consider the situation where we have a collection of $m$ binomial processes
that are all conditionally independent given $\theta$, i.e.
\begin{eqnarray}
X_i\mid n_i,\theta & ~\sim~ & \mathtt{Binom}(n_i, \theta)\,,
\end{eqnarray}
for $i=1,2,\ldots,m$. Letting $\mathbf{X}\doteq(X_1,\ldots,X_m)$
and $\mathbf{n}\doteq(n_1,\ldots,n_m)$, the
joint distribution is given by
\begin{eqnarray}
p(\mathbf{X}\mid\mathbf{n},\theta) & ~=~ &
\prod_{i=1}^m p(X_i\mid n_i,\theta)
\\& = &
\prod_{i=1}^m{n_i \choose X_i}\,\theta^{X_i}(1-\theta)^{n_i-X_i}
\\& = &
\frac{\prod_{i=1}^m n_i!}{\prod_{i=1}^m X_i!\,\prod_{i=1}^m (n_i-X_i)!}\,
\theta^{\sum_{i=1}^m X_i}(1-\theta)^{\sum_{i=1}^m(n_i-X_i)}
\\& = &
\frac{Y!}{\prod_{i=1}^m X_i!}\,\frac{(N-Y)!}{\prod_{i=1}^m(n_i-X_i)!}\,
\frac{\prod_{i=1}^m n_i!}{N!}\times
\frac{N!}{Y!\,(N-Y)!}\,\theta^{Y}(1-\theta)^{N-Y}
\\& = &
%{Y\choose X_1,\ldots,X_m}\,{N-Y\choose n_1-X_1,\ldots,n_m-X_m}
%\left/{N\choose n_1,\ldots,n_m}\right.
{Y\choose\mathbf{X}}\,{N-Y\choose\mathbf{n}-\mathbf{X}}
\left/{N\choose\mathbf{n}}\right.
%\\&&\;\;{}
\times{N\choose Y}\,\theta^{Y}(1-\theta)^{N-Y}
\,,
\end{eqnarray}
where $Y\doteq X_1+\cdots+X_m$ and $N\doteq n_1+\cdots+n_m$.
This is the product of a (special, vector form of) multivariate hypergeometric distribution, which we coin as
\begin{eqnarray}
\mathbf{X}\mid Y & ~\sim~ & \mathtt{VecHyperGeom}(\mathbf{n}, Y)\,,
\end{eqnarray}
with a binomial distribution, namely
\begin{eqnarray}
Y\mid N,\theta & ~\sim~ & \mathtt{Binom}(N, \theta)\,.
\end{eqnarray}
The *vector hypergeometric* may be interpreted in the following way.
Let there be $N$ independent Bernoulli trials, of which $n_1$ are randomly assigned to the first bucket, $n_2$ to the second bucket, et cetera.
Let there be a pre-determined number $Y$ of successes and $N-Y$ failures.
Then the distribution provides the probability that $X_1$ of the successes
were found in the first bucket, $X_2$ in the second bucket, et cetera.

Finally, the total count $Y$ has the marginal distribution
\begin{eqnarray}
p(Y=k\mid\mathbf{n},\theta) & ~=~ &
\underset{\mid X_1+\cdots+X_m=k}{\sum_{X_1=0}^{n_1}\cdots\sum_{X_m=0}^{n_m}}
\,p(\mathbf{X}\mid\mathbf{n},\theta)
\\& = &
{N\choose k}\,\theta^k\,(1-\theta)^{N-k}
~\doteq~ p(Y=k\mid N,\theta)
\,.
\end{eqnarray}
This is just a derivation of the well-known fact that the sum of independent binomially distributed counts is itself binomially distributed.

Now, following the [previous](#Beta-binomial-mixture-distribution
"Section: Beta-binomial mixture distribution") section, we see that the mixture
distribution for $Y$ is a beta-binomial with 
mean $\mu_Y=N\,\mu_\theta$ and variance
\begin{eqnarray}
\sigma^2_Y & ~\doteq~ &
\mathbb{V}[Y\mid\mathbf{n},\alpha,\beta]
~=~ N\mu_\theta(1-\mu_\theta) + N(N-1)\,\sigma^2_\theta\,,
\end{eqnarray}
whereas the mixture distribution for $X_i$ has 
mean $\mu_{X_i}=n_i\,\mu_\theta$ and variance
\begin{eqnarray}
\sigma^2_{X_i} & ~\doteq~ &
\mathbb{V}[X_i\mid\mathbf{n},\alpha,\beta]
~=~ n_i\mu_\theta(1-\mu_\theta) + n_i(n_i-1)\,\sigma^2_\theta\,.
\end{eqnarray}
Consequently, following the reasoning explained in the 
[next](#Negative-Binomial-Distribution
"Section: Negative Binomial Distribution")
section, the excess variance is now given by
\begin{eqnarray}
2\sum_{i=1}^{m-1}\sum_{j=i+1}^m
\mathtt{Cov}[X_i,X_j\mid\mathbf{n},\alpha,\beta]
& ~=~ &
\mathbb{V}\left[\sum_{i=1}^m X_i\mid\mathbf{n},\alpha,\beta\right]
-\sum_{i=1}^m \mathbb{V}\left[X_i\mid\mathbf{n},\alpha,\beta\right]
\\& = &
N^2\,\sigma^2_\theta
-\sum_{i=1}^m n_i^2\,\sigma^2_\theta
\\& = &
2\sum_{i=1}^{m-1}\sum_{j=i+1}^m n_i n_j\,\sigma^2_\theta\,,
\end{eqnarray}
which implies that the covariance between the marginal processeses are given by
\begin{eqnarray}
\Sigma_{ij} & ~\doteq~ &
\mathtt{Cov}[X_i,X_j\mid\mathbf{n},\alpha,\beta]
~=~n_i n_j\,\sigma^2_\theta\,,
\end{eqnarray}
for $i\ne j$, with $\Sigma_{ii}=\sigma^2_{X_i}$. 
Thus, the marginal mixture distributions are not independent.
Lastly, the proportions of the individual processes are given by
\begin{eqnarray}
\theta_i & ~\doteq~ & \frac{n_i}{N}
~=~\frac{\mu_{X_i}}{\mu_Y}
~=~\frac{\sum_{j=1}^m\Sigma_{ij}}{\sum_{i'=1}^m\sum_{j=1}^m\Sigma_{i'j}}
\,.
\end{eqnarray}

## Gamma-Poisson Distribution

Consider a Poisson-like count $X\in\mathbb{Z}^{\ge 0}$ with over-dispersion, namely 
$\mathbb{V}[X]>\mathbb{E}[X]$. This suggests the need for an additonal parameter, beyond the usual Poisson rate $\lambda$, 
to control the extra variance. 
One way in which over-dispersion can arise is when $\lambda$ is only held constant for a single trial 
or a single sequence of trials, but exhibits inter-trial or inter-sequence variation.
This variation may be captured by a prior distribution, giving rise to a continuous mixture distribution.

### Poisson data likelihood

Consider a unit interval of time for which the mean event rate is
$\lambda$. Let the observed number $X$ of events on this interval be distributed as
\begin{eqnarray}
X\mid\lambda & ~\sim~ & \mathtt{Poisson}(\lambda)\,;
\end{eqnarray}
that is
\begin{eqnarray}
p(X=k\mid\lambda) & ~=~ & e^{-\lambda}\,\frac{\lambda^k}{k!}\,,
\end{eqnarray}
for $k=0,1,\ldots,\infty$.
Then the mean and variance are given by
\begin{eqnarray}
\mathbb{E}[X\mid\lambda]~=~\lambda\,, & \;\;\mbox{and}\;\; &
\mathbb{V}[X\mid\lambda]~=~\lambda\,,
\end{eqnarray}
respectively.


### Gamma prior distribution

Consider, in general, some arbitrary prior distribution for the
Poisson event rate $\lambda$. Let this distribution be governed by one or more 
(constant) hyper-parameters denoted by $\boldsymbol{\phi}$, such that
the mean and variance are given by
\begin{eqnarray}
\mu_\lambda ~\doteq~\mathbb{E}[\lambda\mid\boldsymbol{\phi}]\,, 
& ~~~\mbox{and}~~~ & \sigma^2_\lambda~\doteq~\mathbb{V}[\lambda\mid\boldsymbol{\phi}]\,,
\end{eqnarray}
respectively, treating $\lambda$ as a variable.
We are now assuming that $\lambda$ varies for each observation of the count $X$.
Hence, it follows from the [previous](#Poisson-data-likelihood "Section: Poisson data likelihood") section
that the mean and variance of $Y$, without knowledge of $\lambda$, are given
[by](https://en.wikipedia.org/wiki/Law_of_total_expectation
"Wikipedia: Law of total expectation")
\begin{eqnarray}
\mathbb{E}[X\mid\boldsymbol{\phi}] & ~\doteq~ & \mathbb{E}\left[\,\mathbb{E}[X\mid\lambda]\mid\boldsymbol{\phi}\right]
\\& = &
\mathbb{E}[\lambda\mid\boldsymbol{\phi}]~=~\mu_\lambda
\,,
\end{eqnarray}
[and](https://en.wikipedia.org/wiki/Law_of_total_variance
"Wikipedia: Law of total variance")
\begin{eqnarray}
\mathbb{V}[X\mid\boldsymbol{\phi}] & ~\doteq~ & 
\mathbb{V}\left[\,\mathbb{E}[X\mid\lambda]\mid\boldsymbol{\phi}\right]
+
\mathbb{E}\left[\,\mathbb{V}[X\mid\lambda]\mid\boldsymbol{\phi}\right]
\\
& ~=~ & 
\mathbb{V}[\lambda\mid\boldsymbol{\phi}]
+
\mathbb{E}[\lambda\mid\boldsymbol{\phi}]
~=~\sigma^2_\lambda+\mu_\lambda
\,,
\end{eqnarray}
respectively.
Consequently, we observe that $\mathbb{V}[X\mid\boldsymbol{\phi}]>\mathbb{E}[X\mid\boldsymbol{\phi}]$, such that the overall process is over-dispersed compared
to a simple Poisson process. The over-dispersion is explained by the uncertainty about the value of the event rate $\lambda$.

Clearly we require a *proper* prior distribution with finite mean and variance.
Following the rationale discussed [previously](#Beta-prior-distribution "Section: Beta prior distribution"),
we first seek a non-informative prior for the Possion rate $\lambda\in (0,\infty)$ via a variance-stabilising transformation. However, this results in the improper prior
\begin{eqnarray}
p(\lambda) & ~\propto~ & \frac{1}{\sqrt{\mathbb{V}[X\mid\lambda]}}~=~\frac{1}{\sqrt{\lambda}}\,.
\end{eqnarray}

To deduce a suitable form for a proper prior for $\lambda$, consider a sequence $X_1,\ldots,X_m$ of observed counts all sampled with the same value of $\lambda$. The joint likelihood of these Poisson counts is therefore proportional to
$\lambda^{\sum_{i=1}^{m}X_i}\,e^{-m\lambda}$, which has the same form as
the $\texttt{Gamma}(\sum_{i=1}^{m}X_i+1,m)$ distribution.
Consequently, we consider a prior distribution in the general form
\begin{eqnarray}
\lambda & ~\sim~ & \mathtt{Gamma}(\alpha,\beta)\,,
\end{eqnarray}
with hyper-parameters $\boldsymbol{\phi}\doteq(\alpha,\beta)$, where $\alpha$ is the shape parameter and $\beta$ is the rate parameter.
Note that the improper Jeffreys' prior above could notionally be thought of as $\texttt{Gamma}(\frac{1}{2},0)$.

For this gamma distribution, it is 
[known](https://en.wikipedia.org/wiki/Gamma_distribution
"Wikipedia: Gamma distribution")
that the mean and variance are given by
\begin{eqnarray}
\mu_\lambda~=~\frac{\alpha}{\beta}\,, 
& ~~~ & \sigma^2_\lambda~=~\frac{\alpha}{\beta^2}\,,
\end{eqnarray}
respectively.
It then follows from the informal derivation above that
\begin{eqnarray}
\mathbb{E}[X\mid\boldsymbol{\phi}] & ~=~ & \mu_\lambda~=~\frac{\alpha}{\beta}\,,
\end{eqnarray}
and
\begin{eqnarray}
\mathbb{V}[X\mid\boldsymbol{\phi}] & ~=~ & \mu_\lambda+\sigma^2_\lambda
~=~\frac{\alpha\,(\beta+1)}{\beta^2}\,.
\end{eqnarray}
 Hence, the use of the two hyper-parameters
$\alpha$ and $\beta$, compared to just the single rate parameter $\lambda$,
now provides sufficient freedom to flexibly model the mean and variance independently of each other,
thus allowing for over-dispersion.

### Negative binomial distribution

For a single trial
$X\sim\texttt{Poisson}(\lambda)$,
the explicit form of the mixture distribution is derived as
\begin{eqnarray}
p(X=k\mid\boldsymbol{\phi}) & ~=~ & \int_0^\infty p(X=k\mid\lambda)\,p(\lambda\mid\boldsymbol{\phi})\,d\lambda
\\& = &
\int_0^\infty e^{-\lambda}\frac{\lambda^k}{k!}\,
\frac{\beta^\alpha}{\Gamma(\alpha)}\,\lambda^{\alpha-1}\,e^{-\beta\lambda}\,d\lambda
\\& = &
\frac{\beta^\alpha}{k!\,\Gamma(\alpha)}
\int_0^\infty \lambda^{\alpha+k-1}\,e^{-(\beta+1)\lambda}\,d\lambda
%\\& = &
%\frac{\Gamma(\alpha+k)}{k!\,\Gamma(\alpha)}\,\frac{\beta^\alpha}{(\beta+1)^{\alpha+k}}
%\int_0^\infty \frac{(\beta+1)^{\alpha+k}}{\Gamma(\alpha+k)}\,\lambda^{\alpha+k-1}\,e^{-(\beta+1)\lambda}\,d\lambda
\\& = &
\frac{\Gamma(\alpha+k)}{k!\,\Gamma(\alpha)}\,\frac{\beta^\alpha}{(\beta+1)^{\alpha+k}}\,,
\end{eqnarray}
since the final integrand is proportional to the 
$\mathtt{Gamma}(\alpha+k,\beta+1)$
distribution.

To show that this is just the real-valued (*Polya* distribution) version of the
[*negative binomial*](https://en.wikipedia.org/wiki/Negative_binomial_distribution "Wikipedia: Negative binomial distribution")
distribution, we define 
\begin{eqnarray}
\theta & ~\doteq~ & \frac{\beta}{\beta+1}\,,
\end{eqnarray}
to be the probability of a notional *stopping* event (i.e. one out of the required number $\alpha$ of such events),
whereupon
\begin{eqnarray}
p(X=k\mid\alpha,\theta) & ~=~ & 
\frac{\Gamma(\alpha+k)}{k!\,\Gamma(\alpha)}\,\theta^\alpha\,(1-\theta)^k\,.
\end{eqnarray}
Hence, this notionally represents the overall probability of observing a sequence of $k+\alpha$ Bernoulli trials,
where the sequence was terminated as soon as $\alpha$ stopping events occurred.

In the traditional integer-valued (*Pascal* distribution) form, $\alpha>0$ is replaced by $r\in\mathbb{Z}^{>0}$, giving
\begin{eqnarray}
p(X=k\mid r,\theta) & ~=~ & 
\frac{(r+k-1)!}{k!\,(r-1)!}\,\theta^r\,(1-\theta)^k
~=~\binom{k+r-1}{k}\,\theta^r\,(1-\theta)^k\,.
\end{eqnarray}
If we terminate the sequence of trials after just $r=1$ stopping event, then this reduces to the
*geometric* distribution, i.e. 
$\mathtt{Geom}(\theta)\equiv \mathtt{NegBinom}(1,\theta)$.
Furthermore, it can be shown that $X\sim\mathtt{NegBinom}(r,\theta)$ corresponds to the sum of $r$ independent geometric counts, i.e.
$X=Z_1+Z_2+\ldots+Z_r$ with $Z_i\sim\mathtt{Geom}(\theta)$.

### Gamma posterior distribution

For a single observed count $X$, it is easy to deduce the posterior distribution for $\lambda$, as this follows
immediately from the [previous](#Negative-binomial-distribution "Section: Negative binomial distribution") section
via
\begin{eqnarray}
p(\lambda\mid X=k,\boldsymbol{\phi}) & ~=~ & 
\frac{p(X=k\mid\lambda)\,p(\lambda\mid\boldsymbol{\phi})}{p(X=k\mid\boldsymbol{\phi})}
\\& = &
\left.\left[
e^{-\lambda}\frac{\lambda^k}{k!}\,
\frac{\beta^\alpha}{\Gamma(\alpha)}\,\lambda^{\alpha-1}\,e^{-\beta\lambda}
\right]
\right/
\left[
\frac{\Gamma(\alpha+k)}{k!\,\Gamma(\alpha)}\,\frac{\beta^\alpha}{(\beta+1)^{\alpha+k}}
\right]
\\&=&
\frac{(\beta+1)^{\alpha+k}}{\Gamma(\alpha+k)}\,\lambda^{\alpha+k-1}\,e^{-(\beta+1)\lambda}
\,,
\end{eqnarray}
which is exactly
\begin{eqnarray}
\lambda\mid X, \boldsymbol{\phi} & ~\sim~& \texttt{Gamma}(\alpha+X,\beta+1)\,.
\end{eqnarray}
Consequently, the *maximum posterior* estimate is 
[given](https://en.wikipedia.org/wiki/Gamma_distribution "Wikipedia: Gamma distribution") by
\begin{eqnarray}
\hat{\lambda}_\texttt{MP} & ~=~ & \arg\max_{\lambda}p(\lambda\mid X, \boldsymbol{\phi})
~=~\max\left\{0,\; \frac{\alpha+X-1}{\beta+1}\right\}\,.
\end{eqnarray}
Interestingly, the maximum likelihood estimate $\hat{\lambda}_\texttt{ML}=X$ would only be obtained when $\alpha=1$ and $\beta=0$, at which point the prior distribution for
$\lambda$ collapses to zero everywhere.

### Maximum likelihood estimation

In order to derive maximum likehood estimates of the parameters
of the gamma prior, consider a sample $X_1,\ldots,X_N$ of counts, each drawn
independently from the same [negative binomial](#Negative-binomial-distribution "Section: Negative binomial distribution") distribution, namely
\begin{eqnarray}
p(X_i\mid\alpha,\beta) & ~=~ & 
\frac{\Gamma(\alpha+X_i)}{\Gamma(\alpha)\,X_i!}\,
\theta^\alpha\,(1-\theta)^{X_i}
\,,
\end{eqnarray}
where we recall that $\theta=\frac{\beta}{\beta+1}$.
The empirical mean log-likelihood is therefore given by
\begin{eqnarray}
\langle L\rangle(\alpha,\beta) & ~\doteq~ &
\frac{1}{N}\sum_{i=1}^N \ln p(X_i\mid\alpha,\beta)
\\& ~=~ &
\left\langle\ln\Gamma(\alpha+X)-\ln X!\right\rangle
-\ln\Gamma(\alpha)+\langle X\rangle\ln(1-\theta)+\alpha\ln\theta\,,
\end{eqnarray}
where $\langle X\rangle\doteq(X_1+\cdots+X_N)/N$.
Taking the gradient with respect to $\theta$ then gives
\begin{eqnarray}
\nabla_\theta\langle L\rangle & ~=~ &
\frac{\alpha}{\theta}-\frac{\langle X\rangle}{1-\theta}
~=~\frac{\alpha-(\alpha+\langle X\rangle)\,\theta}{\theta\,(1-\theta)}\,,
\end{eqnarray}
such that  the maximum likelihood estimate satisfies
\begin{eqnarray}
\hat{\theta}_\texttt{ML} ~=~ 
\frac{\hat{\alpha}_\texttt{ML}}{\hat{\alpha}_\texttt{ML}
+\langle X\rangle}
& \;\;~\Rightarrow~\;\; &
\hat{\beta}_\texttt{ML}~=~
\frac{\hat{\theta}_\texttt{ML}}{1-\hat{\theta}_\texttt{ML}}
~=~\frac{\hat{\alpha}_\texttt{ML}}{\langle X\rangle}
\,.
\end{eqnarray}

Next, we take the gradient with respect to $\alpha$, which gives
\begin{eqnarray}
\nabla_\alpha\langle L\rangle & ~=~ & 
\langle\psi(\alpha+X)\rangle-\psi(\alpha)+\ln\theta\,.
\end{eqnarray}
Consequently, the maximum likelihood estimate satisfies the nonlinear
equation
\begin{eqnarray}
\left\langle\psi(\hat{\alpha}_\texttt{ML}+X)\right\rangle
-\psi(\hat{\alpha}_\texttt{ML})
+\ln\hat{\theta}_\texttt{ML} & ~=~ & 0\,.
\end{eqnarray}
Now, given the estimate $\hat{\theta}_\texttt{ML}$ above, 
we may define the estimate for $\alpha$ as satisfying 
$f(\hat{\alpha}_\texttt{ML})=0$, where
\begin{eqnarray}
f(\alpha) & ~\doteq~ & 
\langle\psi(\alpha+X)\rangle-\psi(\alpha)
+\ln\alpha-\ln(\alpha+\langle X\rangle)
\\
\Rightarrow
f'(\alpha) & ~=~ & 
\langle\psi'(\alpha+X)\rangle-\psi'(\alpha)
+\frac{\langle X\rangle}{\alpha\,(\alpha+\langle X\rangle)}
\,.
\end{eqnarray}
Consequently, we may then find the maximum likelihood estimate iteratively via 
Newton's method, namely
\begin{eqnarray}
\hat{\alpha}' & ~=~ & \hat{\alpha}-\frac{f(\hat{\alpha})}{f'(\hat{\alpha})}
\,.
\end{eqnarray}


Initial estimates may be obtained using the method of moments. 
It is [known](https://en.wikipedia.org/wiki/Negative_binomial_distribution "Wikipedia: Negative binomial distribution") that
the mean and variance of the Polya distribution are given by
\begin{eqnarray}
\mathbb{E}[X_i\mid\boldsymbol{\phi}]~=~
\frac{\alpha\,(1-\theta)}{\theta}\,, & ~~~\mbox{and}~~~ &
\mathbb{V}[X_i\mid\boldsymbol{\phi}]~=~
\frac{\alpha\,(1-\theta)}{\theta^2}\,,
\end{eqnarray}
respectively. Thus, we equate the distributional mean $\mu_X$ and variance 
$\sigma_X^2$ with the sample mean $\langle X\rangle=\bar{X}$ and variance $s^2_X$, respectively,
to obtain
\begin{eqnarray}
\hat{\theta}~=~\frac{\bar{X}}{s^2_X}\,, & ~~~\mbox{and}~~~ & 
\hat{\alpha}~=~\frac{\bar{X}^2}{s^2_X-\bar{X}}\,.
\end{eqnarray}
Note that the initial estimate $\hat{\alpha}$ will be sensitive to 
$s^2_X-\bar{X}$, which represents the over-dispersion of the negative binomial
distribution in comparison to the underlying Poisson distribution.

### Limiting gamma distribution

As a special case of the 
[negative binomial](#Negative-binomial-distribution "Section: Negative binomial distribution") distribution, we now consider the situation where the probability $\theta$ of a stopping event is small.
Firstly, we consider the *moment generating function* (MGF) of
the [negative binomial](https://en.wikipedia.org/wiki/Negative_binomial_distribution "Wikipedia: Negative binomial distribution"),
namely
\begin{eqnarray}
M_X(t) & ~\doteq~ & \mathbf{E}\left[e^{tX}\mid\alpha,\theta\right]
~=~\left(\frac{\theta}{1-(1-\theta)\,e^t}\right)^\alpha\,,
\end{eqnarray}
which holds for all $t<-\ln(1-\theta)$. Next, we consider the transformation
$Y\doteq\theta X$, which has the MGF
\begin{eqnarray}
M_Y(t) & ~=~ & \mathbb{E}\left[e^{t\theta X}\right]
~=~M_X(\theta t)
~=~
\left(\frac{\theta}{1-(1-\theta)\,e^{\theta t}}\right)^\alpha\,.
\end{eqnarray}
Note that both the numerator and the denominator go to zero as 
$\theta\rightarrow 0$. Hence, we use 
[L'Hôpital's rule](https://en.wikipedia.org/wiki/L%27H%C3%B4pital%27s_rule "Wikipedia: L'Hôpital's rule")
and take the derivatives of each prior to taking the limit, such that
\begin{eqnarray}
\lim_{\theta\rightarrow 0} M_Y(t) & ~=~ & 
\lim_{\theta\rightarrow 0}
\left(\frac{1}{[1-(1-\theta)t]\,e^{\theta t}}\right)^\alpha
~=~(1-t)^{-\alpha}
\,.
\end{eqnarray}
This is just the MGF of the 
[gamma](https://en.wikipedia.org/wiki/Gamma_distribution "Wikipedia: Gamma distribution") distribution for
\begin{eqnarray}
Y & ~=~ & \theta X~\sim~\texttt{Gamma}(\alpha,1)\,,
\end{eqnarray}
from which it follows that $X$ is approximately distributed like
\begin{eqnarray}
X & ~\sim~ & \texttt{Gamma}(\alpha,\theta)\,,
\end{eqnarray}
for small values of $\theta$. Note that the constraint for the gamma MGF
is that $t<\theta$, which is tighter than the corresponding constraint
for the negative binomial MGF since $\theta\le -\ln(1-\theta)$
for all $\theta\in[0,1]$.

## Poisson Sequence Process

Consider a sequence process $\mathbf{X}\doteq(X_1,\ldots,X_n)$ comprised of $n$ independent Poisson sub-processes
with independent rate parameters given by $\boldsymbol{\lambda}\doteq(\lambda_1,\ldots,\lambda_n)$.
The joint distribution is given by
\begin{eqnarray}
p(\mathbf{X}\mid\boldsymbol{\lambda}) & ~=~ & 
\prod_{i=1}^n\frac{e^{-\lambda_i}\,\lambda_i^{X_i}}{X_i!}
~=~
e^{-\rho}\frac{\rho^Y}{Y!}
\times
\frac{Y!}{\prod_{i=1}^n X_i!}\,
\prod_{i=1}^{n}\left(\frac{\lambda_i}{\rho}\right)^{X_i}
\,,
\end{eqnarray}
where we have defined $Y\doteq X_1+\cdots+X_n$
and $\rho\doteq\lambda_1+\cdots+\lambda_n$ for convenience.
We recognize the right-hand term as a
[multinomial](https://en.wikipedia.org/wiki/Multinomial_distribution
"Wikipedia: Multinomial distribution") distribution
with probabilities
\begin{eqnarray}
\boldsymbol{\theta}\doteq (\theta_1,\ldots,\theta_n)~=~
\left(\frac{\lambda_1}{\rho},\ldots,
\frac{\lambda_n}{\rho}\right)
\,,
\end{eqnarray}
and the left-hand term as a Poisson distribution of the total count $Y$ with total event rate
$\rho$.
Formally, $Y$ has the marginal distribution
\begin{eqnarray}
p(Y=k\mid\boldsymbol{\lambda}) & ~=~ &
\underset{X_1+\cdots+X_n=k}{\sum_{X_1=0}^\infty\cdots\sum_{X_n=0}^\infty}
\,p(\mathbf{X}\mid\boldsymbol{\lambda})
~=~
e^{-\rho}\,\frac{\rho^k}{k!}
\,.
\end{eqnarray}
This is just a derivation of the well-known fact that the sum of independent Poission distributed counts is itself Poisson distributed.
We may now reparameterise the sequence process to depend upon both the overall rate $\rho$ and the sub-intervals  $\boldsymbol{\theta}$ of the unit time interval, where $\lambda_i=\theta_i\,\rho$. Consequently, the Poisson sub-processes, which were *unconditionally* independent for arbitrary $\boldsymbol{\lambda}$, are now *conditionally* independent given $\rho$ (which constrains $\boldsymbol{\lambda}$), such that
\begin{eqnarray}
X_i\mid\boldsymbol{\theta},\rho & ~\sim~ & \texttt{Poisson}(\theta_i\,\rho)\,.
\end{eqnarray}

### Marginal correlations within sequences

To see the effect of conditioning the Poisson sub-processes on the overall rate $\rho$, suppose now that $\rho$ has some prior distribution parameterised by $\boldsymbol{\phi}$
(for example, a [gamma](#Gamma-prior-distribution "Section: Gamma prior distribution") distribution), such that
the prior mean and variance are given by $\mu_\rho\doteq\mathbb{E}[\rho\mid\boldsymbol{\phi}]$ and 
$\sigma^2_\rho\doteq\mathbb{V}[\rho\mid\boldsymbol{\phi}]$, respectively. Then we deduce that the conditional distributions of the sub-processes,
in the absence of knowing $\rho$, and keeping $\boldsymbol{\theta}$ constant,
have means and variances given by
\begin{eqnarray}
\mu_{X_i} & ~\doteq~ &
\mathbb{E}[X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}] ~=~
\mathbb{E}\left[
\mathbb{E}[X_i\mid\boldsymbol{\theta},\rho]
\mid\boldsymbol{\theta},\boldsymbol{\phi}\right]
\\& = &
\mathbb{E}[\theta_i\,\rho\mid\boldsymbol{\theta},\boldsymbol{\phi}]
~=~\theta_i\,\mu_\rho\,,
\end{eqnarray}
and
\begin{eqnarray}
\sigma^2_{X_i} & ~\doteq~ &
\mathbb{V}[X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}] ~=~
\mathbb{E}\left[
\mathbb{V}[X_i\mid\boldsymbol{\theta},\rho]
\mid\boldsymbol{\theta},\boldsymbol{\phi}\right]
+
\mathbb{V}\left[
\mathbb{E}[X_i\mid\boldsymbol{\theta},\rho]
\mid\boldsymbol{\theta},\boldsymbol{\phi}\right]
\\& = &
\mathbb{E}[\theta_i\,\rho\mid\boldsymbol{\theta},\boldsymbol{\phi}]
+
\mathbb{V}[\theta_i\,\rho\mid\boldsymbol{\theta},\boldsymbol{\phi}]
~=~
\theta_i\,\mu_\rho+\theta_i^2\,\sigma^2_\rho\,,
\end{eqnarray}
respectively.
Consequently, these conditional distributions are over-dispersed. Similarly, we see for any two distinct sub-processes, say $i\ne j$, that
\begin{eqnarray}
\mathbb{E}[X_i\,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}] & ~=~ &
\mathbb{E}\left[
\mathbb{E}[X_i\,X_j\mid\boldsymbol{\theta},\rho]
\mid\boldsymbol{\theta},\boldsymbol{\phi}\right]
\\& = &
\mathbb{E}[\theta_i\,\rho\;\theta_j\,\rho\mid\boldsymbol{\theta},\boldsymbol{\phi}]
~=~\theta_i\,\theta_j\,(\sigma^2_\rho+\mu_\rho^2)\,,
\end{eqnarray}
such that
\begin{eqnarray}
\texttt{Cov}[X_i,\,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}] & ~=~ &
\mathbb{E}[X_i\,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}] - 
\mathbb{E}[X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}]\;
\mathbb{E}[X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}]
~=~\theta_i\,\theta_j\,\sigma^2_\rho\,.
\end{eqnarray}
Hence, the conditional distributions change from being uncorrelated, when $\rho$ and $\boldsymbol{\theta}$ are held constant, to being correlated
when $\rho$ is varied before each process sample.
Note that these correlations are not due to *intrinsic serial* correlations between successive sub-processes, since we assumed that the sub-processes
were conditionally independent given all requisite parameters. Instead, the correlations arise due to the fact that 
the rates $\boldsymbol{\lambda}$ are now constrained to sum to $\rho$, even if $\rho$ is unknown. 
In other words, the correlations are due to the uncertainties arising from having only imperfect information about the parameter $\rho$
that ties the sub-processes together.

To see the effect of these correlations on the conditional distribution of the total count $Y$, observe that
its mean is given by
\begin{eqnarray}
\mu_{Y} & ~\doteq~ &
\mathbb{E}[Y\mid\boldsymbol{\theta},\boldsymbol{\phi}] ~=~
\mathbb{E}\left[\left.\sum_{i=1}^{n}X_i\;\right|\;\boldsymbol{\theta},\boldsymbol{\phi}\right]
%~=~\sum_{i=1}^{n}\theta_i\,\mu_\rho
~=~\mu_\rho\,,
\end{eqnarray}
since the multinomial probabilities $\boldsymbol{\theta}$ must sum to unity.
Similarly, the conditional variance is given by
\begin{eqnarray}
\sigma^2_{Y} & ~\doteq~ &
\mathbb{V}[Y\mid\boldsymbol{\theta},\boldsymbol{\phi}]
~=~
\mathbb{V}\left[\left.\sum_{i=1}^{n}X_i\;\right|\;\boldsymbol{\theta},\boldsymbol{\phi}\right]
\\& = & 
\sum_{i=1}^n\sum_{j=1}^n\mathtt{Cov}[X_i,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}]
\\& = &
\sum_{i=1}^{n}\left\{\mathbb{V}[X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}] +
\underset{j\ne i}{\sum_{j=1}^n}
\mathtt{Cov}[X_i,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}]\right\}
\\
& ~=~ &
\sum_{i=1}^{n}\left\{\theta_i\,\mu_\rho+\theta_i^2\,\sigma^2_\rho+\theta_i\,(1-\theta_i)\,\sigma^2_\rho\right\}
~=~\sigma^2_\rho+\mu_\rho\,.
\end{eqnarray}
We therefore deduce from this derivation that a clear consequence of sub-process correlation is that
\begin{eqnarray}
\mathbb{V}[Y\mid\boldsymbol{\theta},\boldsymbol{\phi}]
& ~>~ &
\sum_{i=1}^{n}\mathbb{V}[X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}]\,,
\\
\Rightarrow \sigma^2_\rho & ~=~ & 
\mathbb{V}[Y\mid\boldsymbol{\theta},\boldsymbol{\phi}] - \mathbb{E}[Y\mid\boldsymbol{\theta},\boldsymbol{\phi}]
\\\nonumber
& > & \sum_{i=1}^{n}\left\{
\mathbb{V}[X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}] - \mathbb{E}[X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}]
\right\}~=~\sum_{i=1}^{n}\theta_i^2\,\sigma^2_\rho\,,
\end{eqnarray}
i.e. the conditional distribution for $Y$ must be even more over-dispersed than suggested by the sum of over-dispersions of the individual sub-processes.

We may also make several further observations. Firstly, note that both the conditional mean 
$\mu_{Y}$ and variance
$\sigma^2_{Y}$ are actually independent of
the $\boldsymbol{\theta}$ proportions. Consequently, these quantities also give the unconditional mean $\mu_Y$ and variance $\sigma^2_Y$,
respectively, in the absence of knowing either $\rho$ or $\boldsymbol{\theta}$.
In contrast, the conditional mean $\mu_{X_i}$
and variance $\sigma^2_{X_i}$
of the sub-process for $X_i$ both depend strongly upon $\boldsymbol{\theta}$.

Secondly, note that we may construct the conditional covariance matrix 
$\boldsymbol{\Sigma}_{\mathbf{X}}$ for the sequence process $\mathbf{X}$,
where the $ij$-th element of the matrix is given by 
\begin{eqnarray}
\Sigma_{ij} & ~\doteq~ & \mathtt{Cov}[X_i,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}]\,.
\end{eqnarray}
It follows from the derivation above that the variance $\sigma^2_Y$ is given by the sum of all elements of $\boldsymbol{\Sigma}_{\mathbf{X}}$.
Similarly, the conditional covariance between any $X_i$ and $Y$ is given by the sum of the $i$-th row of $\boldsymbol{\Sigma}_{\mathbf{X}\mid\boldsymbol{\theta}}$, since
\begin{eqnarray}
\mathtt{Cov}[X_i,Y\mid\boldsymbol{\theta},\boldsymbol{\phi}] & ~=~ &
\sum_{j=1}^n\mathtt{Cov}[X_i,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}]\,.
\end{eqnarray}
Consequently, the expected proportion $\theta_i$ of $Y$ allocated to $X_i$ has the property that
\begin{eqnarray}
\theta_i & ~=~ &
\frac{\mu_{X_i}}{\mu_Y}~=~
\frac{\sum_{j=1}^n\Sigma_{ij}}
{\sigma^2_Y}\,,
\end{eqnarray}
for $i=1,2,\ldots,n$.

### Gamma-Poisson sequences

Recall from the 
[introductory](#Poisson-sequence-process "Section: Poisson sequence process") section that
the joint distribution for the sequence $\mathbf{X}$
may be written as the product of a Poisson distribution in the total count 
$Y\doteq X_1+\cdots+X_n$ and a multinomial 
distribution of $\mathbf{X}$ conditional on $Y$.
We now follow the suggestion of the 
[previous](#Marginal-correlations-within-sequences "Section: Marginal correlations within sequences")
section and let the Posson rate $\rho$ be sampled from a prior Gamma distribution, viz
\begin{eqnarray}
\rho\mid\boldsymbol{\phi} & ~\sim~ & \mathtt{Gamma}(\alpha,\beta)\,.
\end{eqnarray}
As shown [previously](#Negative-binomial-distribution "Section: Negative binomial distribution"),
this turns the Poisson distribution for $Y$ into a 
[negative binomial](https://en.wikipedia.org/wiki/Negative_binomial_distribution 
"Wikipedia: Negative binomial distribution") distribution.
For convenience, we write the latter in an alternative parameterisation using 
$\boldsymbol{\phi}\doteq(\alpha,\beta)$, namely
\begin{eqnarray}
Y\mid\boldsymbol{\phi} & ~\sim~ & \texttt{NegBinom}(\alpha,\beta)\,,
\end{eqnarray}
such that
\begin{eqnarray}
p(Y\mid\boldsymbol{\phi}) & ~=~ & 
\frac{\Gamma(\alpha+Y)}{Y!\,\Gamma(\alpha)}\,
\left(\frac{\beta}{\beta+1}\right)^\alpha\,
\left(\frac{1}{\beta+1}\right)^{Y}\,.
\end{eqnarray}
Consequently, the joint distribution of the sequence $\mathbf{X}$, in the absence of knowing
the true rate $\rho$, is the product of the multinomial distribution and the
negative binomial distribution, namely
\begin{eqnarray}
p(\mathbf{X}\mid\boldsymbol{\theta},\boldsymbol{\phi}) & ~=~ & 
\frac{\Gamma\left(\alpha+\sum_{i=1}^{n}X_i\right)}{\Gamma(\alpha)\,\prod_{i=1}^{n}X_i!}\,
\left(\frac{\beta}{\beta+1}\right)^\alpha\,
\prod_{i=1}^{n}\left(\frac{\theta_i}{\beta+1}\right)^{X_i}\,.
\end{eqnarray}

What now is the marginal distribution of any $X_i$ for $i=1,\ldots,n$?
We could, with care, obtain $p(X_i\mid\boldsymbol{\theta},\boldsymbol{\phi})$
by direct marginalisation of $p(\mathbf{X}\mid\boldsymbol{\theta},\boldsymbol{\phi})$.
Alternatively, we could derive the former from first principles.
However, in is instructive to consider a third alternative.
Recall that $X_i$ has a conditional Poisson distribution,  namely
\begin{eqnarray}
X_i\mid\boldsymbol{\theta},\rho & ~\sim~ & \mathtt{Poisson}(\theta_i\rho)\,.
\end{eqnarray}
Now, since $\rho$ is gamma distributed, it
[follows](https://en.wikipedia.org/wiki/Gamma_distribution "Wikipedia: Gamma distribution")
that
\begin{eqnarray}
\rho\mid\boldsymbol{\phi}~\sim~\mathtt{Gamma}(\alpha,\beta)
& ~~\Rightarrow~~ &
\theta_i\rho\mid\boldsymbol{\theta},\boldsymbol{\phi}~\sim~
\mathtt{Gamma}\left(\alpha,\frac{\beta}{\theta_i}\right)
\,.
\end{eqnarray}
Consequently, the marginal distribution for $X_i$ must be a negative binomial, i.e.
\begin{eqnarray}
X_i\mid\boldsymbol{\theta},\boldsymbol{\phi} & ~\sim~ & 
\mathtt{NegBinom}\left(\alpha,\frac{\beta}{\theta_i}\right)\,,
\end{eqnarray}
such that
\begin{eqnarray}
p(X_i\mid\boldsymbol{\theta},\boldsymbol{\phi}) & ~=~ & 
\frac{\Gamma(\alpha+X_i)}{\Gamma(\alpha)\,X_i!}\,
\left(\frac{\beta}{\beta+\theta_i}\right)^\alpha\,
\left(\frac{\theta_i}{\beta+\theta_i}\right)^{X_i}\,.
\end{eqnarray}
We now see why direct marginalisation
is often difficult unless one is an expert. In this case,
note that the original $\beta+1$ denominator was actually derived as
$\beta+\theta_1+\cdots+\theta_n$; hence, it is replaced by $\beta+\theta_i$ in
the marginalisation. 

The marginal, joint distribution of $X_i$ and $X_j$, for $j\ne i$, may now be
derived from either direct marginalisation or first principles;
both approaches lead to
\begin{eqnarray}
p(X_i,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi}) & ~=~ & 
\frac{\Gamma(\alpha+X_i+X_j)}{\Gamma(\alpha)\,X_i!\,X_j!}\,
\left(\frac{\beta}{\beta+\theta_i+\theta_j}\right)^\alpha\,
\left(\frac{\theta_i}{\beta+\theta_i+\theta_j}\right)^{X_i}\,
\left(\frac{\theta_j}{\beta+\theta_i+\theta_j}\right)^{X_j}\,.
\end{eqnarray}
Finally, the corresponding predictive distribution of $X_j$ conditional on $X_i$ is given by
\begin{eqnarray}
p(X_j\mid X_i,\boldsymbol{\theta},\boldsymbol{\phi}) & ~=~ & 
\frac{p(X_i,X_j\mid\boldsymbol{\theta},\boldsymbol{\phi})}
{p(X_i\mid\boldsymbol{\theta},\boldsymbol{\phi})}
\\& = &
\frac{\Gamma(\alpha+X_i+X_j)}{\Gamma(\alpha+X_i)\,X_j!}\,
\left(\frac{\beta+\theta_i}{\beta+\theta_i+\theta_j}\right)^{\alpha+X_i}\,
\left(\frac{\theta_j}{\beta+\theta_i+\theta_j}\right)^{X_j}\,.
\end{eqnarray}
Observe that this is exactly
\begin{eqnarray}
X_j\mid X_i,\boldsymbol{\theta},\boldsymbol{\phi}
& ~\sim~ & \mathtt{NegBinom}\left(\alpha+X_i,\frac{\beta+\theta_i}{\theta_j}\right)\,.
\end{eqnarray}
We therefore deduce that sequential predictions are distributed as
\begin{eqnarray}
X_k\mid\mathbf{X}_{1:k-1},\boldsymbol{\theta},\boldsymbol{\phi}
& ~\sim~ & \mathtt{NegBinom}\left(\alpha+\sum_{i=1}^{k-1}X_i,
\frac{\beta+\sum_{i=1}^{k-1}\theta_i}{\theta_k}\right)\,,
\end{eqnarray}
where we have defined the partially observed sub-sequence
$\mathbf{X}_{1:k}\doteq(X_1,\ldots,X_k)$ for $k=1,\ldots,n$.

## Gamma Sequence Process

Consider a sequence process $\mathbf{X}\doteq(X_1,\ldots,X_n)$ comprised of $n$ conditionally independent 
[gamma](https://en.wikipedia.org/wiki/Gamma_distribution "Wikipedia: Gamma distribution")-distributed 
sub-processes
that all share a common rate parameter $\beta$, and have independent shape parameters given by $\boldsymbol{\alpha}\doteq(\alpha_1,\ldots,\alpha_n)$.
Formally, the $i$-th sub-process has the distribution 
\begin{eqnarray}
X_i\mid\boldsymbol{\alpha},\beta & ~\sim~ & \texttt{Gamma}(\alpha_i, \beta)\,,
\end{eqnarray}
for $i=1,\ldots,n$.
The joint distribution of the sequence is then given by
\begin{eqnarray}
p(\mathbf{X}\mid\boldsymbol{\alpha},\beta) & ~=~ & 
\prod_{i=1}^n\frac{\beta^{\alpha_i}\,X_i^{\alpha_i-1}\,e^{-\beta X_i}}{\Gamma(\alpha_i)}
\\ & ~=~ &
Y^{-(n-1)}\times
\frac{\beta^{\alpha_0}\,Y^{\alpha_0-1}\,e^{-\beta Y}}{\Gamma(\alpha_0)}
\times
\frac{\Gamma(\alpha_0)}{\prod_{i=1}^n\Gamma(\alpha_i)}\,
\prod_{i=1}^n\left(\frac{X_i}{Y}\right)^{\alpha_i-1}\,
\,,
\end{eqnarray}
where we have defined $Y\doteq X_1+\cdots+X_n$
and $\alpha_0\doteq\alpha_1+\cdots+\alpha_n$ for convenience.
The second term in the joint density is just the gamma density for the total $Y$, i.e.
\begin{eqnarray}
Y\mid\alpha_0,\beta & ~\sim~ & \texttt{Gamma}(\alpha_0,\beta)\,.
\end{eqnarray}
Similarly, the third term is just the [Dirichlet](https://en.wikipedia.org/wiki/Dirichlet_distribution "Wikipedia: Dirichlet distribution") 
density for the proportions
\begin{eqnarray}
\boldsymbol{\theta} & ~\doteq~ & \left(\frac{X_1}{Y},\ldots,\frac{X_n}{Y}\right)~=~(\theta_1,\ldots,\theta_n)\,,
\end{eqnarray}
i.e.
\begin{eqnarray}
\boldsymbol{\theta}\mid\boldsymbol{\alpha} & ~\sim~ & \texttt{Dirich}(\boldsymbol{\alpha})\,.
\end{eqnarray}
Finally, the first term is just the Jacobian of the transformation from $(X_1,\ldots,X_n)$ to
$(Y,\theta_1,\ldots,\theta_{n-1})$ using the fact that $\theta_n=1-\sum_{i=1}^{n-1}\theta_i$.
Consequently, we have derived the well known fact that the sum of conditionally independent gamma variables (that share the same
rate parameter $\beta$) is also a gamma-distributed variable.
Interestingly, we also see that drawing from a Dirichlet distribution is equivalent to renormalising a sequence of draws from independent gamma distributions (e.g. with $\beta=1$).

### Over-dispersed gamma distribution

Given the parameters $\boldsymbol{\alpha}$ and $\beta$, the variance in $Y$ is exactly accounted for by the variances
of the individual $X_i$ variables.
To see this, note that the mean and variance for each $X_i$ are given by
\begin{eqnarray}
\mu_{X_i}~\doteq~\mathbb{E}[X_i\mid\boldsymbol{\alpha},\beta]~=~\frac{\alpha_i}{\beta}\,,
& ~~~ & 
\sigma^2_{X_i}~\doteq~\mathbb{V}[X_i\mid\boldsymbol{\alpha},\beta]~=~\frac{\alpha_i}{\beta^2}\,,
\end{eqnarray}
respectively, for $i=1,\ldots,n$. Similarly, for $Y$ the mean and variance are given by
\begin{eqnarray}
\mu_{Y}~\doteq~\mathbb{E}[Y\mid\boldsymbol{\alpha},\beta]~=~\frac{\alpha_0}{\beta}\,,
& ~~~ & 
\sigma^2_{Y}~\doteq~\mathbb{V}[Y\mid\boldsymbol{\alpha},\beta]~=~\frac{\alpha_0}{\beta^2}\,,
\end{eqnarray}
respectively. Consequently, we see that
\begin{eqnarray}
\mathbb{V}[Y\mid\boldsymbol{\alpha},\beta] & ~=~ & 
\sum_{i=1}^n \mathbb{V}[X_i\mid\boldsymbol{\alpha},\beta]
\,,
\end{eqnarray}
i.e. there is no room for extra dispersion in $Y$ if $\boldsymbol{\alpha}$ and $\beta$ are both held constant.

It turns out that letting $\boldsymbol{\alpha}$ vary does not provide any surplus variance to $Y$, since 
$\sigma^2_{\alpha_0}=\sigma^2_{\alpha_1}+\cdots+\sigma^2_{\alpha_n}$ due to the independence of each $\alpha_i$,
leading to
\begin{eqnarray}
\mathbb{V}[Y\mid\beta] & ~=~ & \sum_{i=1}^n \mathbb{V}[X_i\mid\beta]\,.
\end{eqnarray}
Consequently, we investigate letting $\beta$ vary before each sampling of $\mathbf{X}$, by assuming a draw from some arbitrary prior distribution with parameters $\boldsymbol{\phi}$. 
To make the mathematics more tractable, we first convert the rate parameter $\beta$ into a corresponding scale parameter $\gamma\doteq\frac{1}{\beta}$.
The mean value of $X_i$ is now given by
\begin{eqnarray}
\mathbb{E}[X_i\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ & 
\mathbb{E}[\,\mathbb{E}[X_i\mid\boldsymbol{\alpha},\beta]\mid\boldsymbol{\alpha},\boldsymbol{\phi}\,]
~=~\mathbb{E}[\alpha_i\,\gamma\mid\boldsymbol{\alpha},\boldsymbol{\phi}]
~=~\alpha_i\,\mu_{\gamma}
\,.
\end{eqnarray}
Similarly, the variance of $X_i$ is now given by
\begin{eqnarray}
\mathbb{V}[X_i\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ & 
\mathbb{V}[\,\mathbb{E}[X_i\mid\boldsymbol{\alpha},\beta]\mid\boldsymbol{\alpha},\boldsymbol{\phi}\,]
+
\mathbb{E}[\,\mathbb{V}[X_i\mid\boldsymbol{\alpha},\beta]\mid\boldsymbol{\alpha},\boldsymbol{\phi}\,]
\\& ~=~ &
\mathbb{V}[\alpha_i\,\gamma\mid\boldsymbol{\alpha},\boldsymbol{\phi}]
+
\mathbb{E}[\alpha_i\,\gamma^2\mid\boldsymbol{\alpha},\boldsymbol{\phi}]
\\& ~=~ &
\alpha_i^2\,\sigma^2_\gamma+\alpha_i\,(\sigma^2_\gamma+\mu_\gamma^2)
~=~\alpha_i\,(\alpha_i+1)\,\sigma^2_\gamma+\alpha_i\,\mu_\gamma^2
\,.
\end{eqnarray}
Following a similar analysis for $Y$, we obtain its corresponding mean and variance to be given by
\begin{eqnarray}
\mathbb{E}[Y\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ & \alpha_0\,\mu_\gamma\,,
\end{eqnarray}
and
\begin{eqnarray}
\mathbb{V}[Y\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ & \alpha_0\,(\alpha_0+1)\,\sigma^2_\gamma+\alpha_0\,\mu_\gamma^2\,,
\end{eqnarray}
respectively.
Consequently, the extra dispersion is given by
\begin{eqnarray}
\mathbb{V}[Y\mid\boldsymbol{\alpha},\boldsymbol{\phi}] - \sum_{i=1}^n \mathbb{V}[X_i\mid\boldsymbol{\alpha},\boldsymbol{\phi}]
& ~=~ & \left(\alpha_0^2-\sum_{i=1}^n\alpha_i^2\right)\,\sigma^2_\gamma~>~0
\,.
\end{eqnarray}

As a special case, we may consider the situation where the sub-processes all have exactly the same distribution, i.e. $\alpha_i\doteq\bar{\alpha}$ and $\alpha_0\doteq n\bar{\alpha}$. In this case, we may pool the observations $\mathbf{X}=(X_1,\ldots,X_n)$ to form a single observable variable, say $X$.
The key equations are then
\begin{eqnarray}
\mu_X~=~\bar{\alpha}\,\mu_\gamma\,, 
& ~~~\sigma^2_X~=~\bar{\alpha}\,(\bar{\alpha}+1)\,\sigma^2_\gamma+\bar{\alpha}\,\mu_\gamma^2\,,
& ~~~\sigma^2_Y~=~n\bar{\alpha}\,(n\bar{\alpha}+1)\,\sigma^2_\gamma+n\bar{\alpha}\,\mu_\gamma^2\,,
\end{eqnarray}
which can to be solved in terms of the unknowns, namely $\bar{\alpha}$, $\mu_\gamma$ and $\sigma^2_\gamma$.

### Gamma-gamma distribution

One [recommended](https://en.wikipedia.org/wiki/Gamma_distribution "Wikipedia: Gamma distribution") 
prior distribution for the rate parameter $\beta$ is another gamma distribution.
Hence, consider hyper-parameters $\boldsymbol{\phi}\doteq(\kappa,\lambda)$ such that
\begin{eqnarray}
\beta\mid\boldsymbol{\phi} & ~\sim~ & \texttt{Gamma}(\kappa, \lambda)\,.
\end{eqnarray}
The gamma-gamma mixture distribution then takes the form
\begin{eqnarray}
p(Y\mid\boldsymbol{\alpha},\boldsymbol{\phi}) & ~\doteq~ & 
\int_0^\infty p(Y\mid\boldsymbol{\alpha},\beta)\,p(\beta\mid\boldsymbol{\phi})\,d\beta
\\& ~=~ &
\int_0^\infty
\frac{\beta^{\alpha_0}\,Y^{\alpha_0-1}\,e^{-\beta Y}}{\Gamma(\alpha_0)}\,
\frac{\lambda^{\kappa}\,\beta^{\kappa-1}\,e^{-\lambda\beta}}{\Gamma(\kappa)}\,
d\beta
\\& ~=~ &
\frac{Y^{\alpha_0-1}\,\lambda^{\kappa}}{\Gamma(\alpha_0)\,\Gamma(\kappa)}
\int_0^\infty
\beta^{\kappa+\alpha_0-1}\,e^{-(\lambda+Y)\beta}\,
d\beta
\\& ~=~ &
\frac{Y^{\alpha_0-1}\,\lambda^{\kappa}}{\Gamma(\alpha_0)\,\Gamma(\kappa)}\,
\frac{\Gamma(\kappa+\alpha_0)}{(\lambda+Y)^{\kappa+\alpha_0}}
\\& ~=~ &
\frac{1}{\lambda}\,\frac{\Gamma(\kappa+\alpha_0)}{\Gamma(\alpha_0)\,\Gamma(\kappa)}\,
\left(\frac{Y}{Y+\lambda}\right)^{\alpha_0-1}\,
\left(\frac{\lambda}{Y+\lambda}\right)^{\kappa+1}\,
\,,
\end{eqnarray}
since the final integrand is proportional to a $\texttt{Gamma}(\kappa+\alpha_0,\lambda+Y)$ density.

It is now tempting to define a probability $\tau$ that obeys the transformation
\begin{eqnarray}
\tau ~\doteq~\frac{Y}{Y+\lambda} & ~~~\Leftrightarrow~~~ &
Y~=~\frac{\lambda\tau}{1-\tau}\,.
\end{eqnarray}
The Jacobian is then given by
\begin{eqnarray}
\frac{\partial\tau}{\partial Y} & ~=~ & \tau\,\frac{\partial\ln\tau}{\partial Y}
~=~\tau\frac{\partial}{\partial Y}\left[\ln Y-\ln(Y+\lambda)
\right]
\nonumber\\
& ~=~ & \frac{\tau}{Y}-\frac{\tau}{Y+\lambda}
~=~\frac{1-\tau}{\lambda}-\frac{\tau\,(1-\tau)}{\lambda}
\nonumber\\
& ~=~ & \frac{(1-\tau)^2}{\lambda}\,,
\end{eqnarray}
such that the density of $\tau$ is given by
\begin{eqnarray}
p(\tau\mid\boldsymbol{\alpha},\boldsymbol{\phi}) & ~=~ &
p(Y\mid\boldsymbol{\alpha},\boldsymbol{\phi})\,\left(\frac{\partial\tau}{\partial Y}\right)^{-1}
~=~
\frac{\Gamma(\kappa+\alpha_0)}{\Gamma(\alpha_0)\,\Gamma(\kappa)}\,
\tau^{\alpha_0-1}\,
(1-\tau)^{\kappa-1}\,,
\end{eqnarray}
i.e.
\begin{eqnarray}
\tau & ~\sim~ & \texttt{Beta}(\alpha_0,\kappa)\,.
\end{eqnarray}
Observe that the distribution of $\tau$ does not  depend upon $\lambda$.
Also note that $\lambda$ is both a rate parameter for $\beta$ and a scale parameter for $Y$.

It then follows that
\begin{eqnarray}
\mathbb{E}[Y^m\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ &
\int_{0}^{1}\left(\frac{\lambda\tau}{1-\tau}\right)^m\,
\frac{\Gamma(\kappa+\alpha_0)}{\Gamma(\alpha_0)\,\Gamma(\kappa)}\,
\tau^{\alpha_0-1}\,(1-\tau)^{\kappa-1}\,d\tau
\nonumber\\
& ~=~ &
\lambda^m\,\frac{\Gamma(\kappa+\alpha_0)}{\Gamma(\alpha_0)\,\Gamma(\kappa)}\,
\int_{0}^{1}\tau^{\alpha_0+m-1}\,(1-\tau)^{\kappa-m-1}\,d\tau
\nonumber\\
& ~=~ &
\frac{\lambda^m\,\Gamma(\alpha_0+m)\,\Gamma(\kappa-m)}{\Gamma(\alpha_0)\,\Gamma(\kappa)}
\,.
\end{eqnarray}
For $m=1$, this gives
\begin{eqnarray}
\mathbb{E}[Y\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ & \frac{\lambda\,\alpha_0}{\kappa-1}\,,
\end{eqnarray}
recalling that $\Gamma(z+1)=z\,\Gamma(z)$. Similarly, for $m=2$, we obtain
\begin{eqnarray}
\mathbb{E}[Y^2\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ & 
\frac{\lambda^2\,\alpha_0\,(\alpha_0+1)}{(\kappa-1)\,(\kappa-2)}\,,
\end{eqnarray}
from which it follows that $\kappa>2$. Hence, the variance of $Y$ is given by
\begin{eqnarray}
\mathbb{V}[Y\mid\boldsymbol{\alpha},\boldsymbol{\phi}] & ~=~ & 
\mathbb{E}[Y^2\mid\boldsymbol{\alpha},\boldsymbol{\phi}] -
\mathbb{E}[Y\mid\boldsymbol{\alpha},\boldsymbol{\phi}]^2
\nonumber\\
& ~=~ & \frac{\lambda^2\,\alpha_0}{\kappa-1}\,\left[
\frac{\alpha_0+1}{\kappa-2}-\frac{\alpha_0}{\kappa-1}
\right]\,.
\end{eqnarray}
Consequently, a similar analysis gives rise to the mean and variance of each $X_i$ as
\begin{eqnarray}
\mathbb{E}[X_i\mid\boldsymbol{\alpha},\boldsymbol{\phi}] 
& ~=~ & \frac{\lambda\,\alpha_i}{\kappa-1}\,,
\end{eqnarray}
and
\begin{eqnarray}
\mathbb{V}[X_i\mid\boldsymbol{\alpha},\boldsymbol{\phi}]
& ~=~ & \frac{\lambda^2\,\alpha_i}{\kappa-1}\,\left[
\frac{\alpha_i+1}{\kappa-2}-\frac{\alpha_i}{\kappa-1}
\right]\,,
\end{eqnarray}
respectively, for $i=1,2,\ldots,n$.
