# Appendix E: Distributional Invariance

In this appendix, we describe and summarise two methodologies for deriving continuous probability distributions based on specified invariance properties. In practice, such an invariant distribution might be useful as an *ignorance* prior, when little is known in advance about the distributional properties.

The first approach we analyse in depth is that of *transformation group invariance*, which was championed by
Jaynes [[1]](#References "Reference: Prior probabilities and transformation groups"). 
The second approach is that of *minimum information* priors,
as discussed by
Zellner [[5]](#References "Reference: An introduction to Bayesian inference in econometrics").

Jaynes' basic proposition was that if two rational agents both start with exactly the same information, then for consistency their subjective distributions for a process of common interest must be identical.
A problem arises, however, when the two agents observe the same process from different viewpoints. To agree on the same distribution, that distribution must be invariant to their viewpoints. To put this another way, the distribution
should be invariant to a specified group of continuous transformations of the viewpoint of any observer, subject to relevant constraints.

In this appendix, we examine the theory of transformational invariance
and derive a general identity that the distribution must obey.
We the proceed to demonstrate some practical examples by which actual
distributions are derived via different groups of transformations.
As an alternative approach, we also briefly examine minimum information priors,
along with a few examples of their use. However, we conclude that
minimum information priors are not invariant to transformations.

The major sections are:

## Transformation Group Invariance

As noted briefly above, Jaynes' idea was that if there is a continous transformation between the viewpoints of two observers, and further that if knowledge of the transformation does not make the prior knowledge of the observers differ, then they must both, for consistency, assign *exactly* the same prior distribution. 

However, Jaynes himself offered an example that was an exception to this rule, namely
the case of a contraction mapping from a circle to a smaller, concentric circle. In this case, Jaynes rightly states that the distribution conditional on the inner circle is not equal to, but instead *proportional* to, the distribution conditional on the outer circle, with the constant of proportionality being the ratio of the factors needed to normalise the respective proper distributions.

In fact, the need for proportionality was also demonstrated by
Milne [[2]](#References "Reference: A note on scale invariance") 
for improper distributions resulting from power-law transformations
(as discussed later in the 
[example](#Nonlinear-rescaling-plus-relocation "Section: Nonlinear rescaling plus relocation") of nonlinear scale invariance).

Consequently, we aim to derive the general invariance relations, including proportionality, from first principles. It should be noted, however, that all of the examples for which Jaynes did *not* include proportionality actually result in a proportionality constant of unity, and hence remain correct.

### Transformation geometry

We begin with two continuous spaces, $\mathcal{U}\subseteq\mathbb{R}^n$
and $\mathcal{V}\subseteq\mathbb{R}^m$, and a continuous transformation,
$\mathbf{h}: \mathcal{U}\mapsto\mathcal{V}$,
such that the point $\mathbf{u}=(u_1,\ldots,u_n)\in\mathcal{U}$ is mapped to its counterpart
$\mathbf{v}=(v_1,\ldots,v_m)\in\mathcal{V}$ via $\mathbf{v}=\mathbf{h}(\mathbf{u})$.

We further suppose that $\mathbf{h}$ is differentiable, such that an infinitesimal 
displacement from $\mathbf{u}$, denoted by 
$d\mathbf{u}\doteq(du_1,\ldots,du_n)$, is mapped to its counterpart displacement
$d\mathbf{v}\doteq(dv_1,\ldots,dv_n)$ from $\mathbf{v}$ via
\begin{eqnarray}
d\mathbf{v} & = & 
\frac{\partial\mathbf{h}}{\partial\mathbf{u}^T}\,d\mathbf{u}\,.
\end{eqnarray}

Similarly, we suppose that the infinitesimal volume element 
$dU=|d\mathbf{u}|\doteq du_1\cdots du_n$ about $\mathbf{u}$ is mapped to its 
counterpart volume element
$dV=|d\mathbf{v}|\doteq dv_1\cdots dv_m$ about $\mathbf{v}$. 
At this juncture, we further require
$\mathbf{h}$ to be invertible, such that no information is lost in the transformation. Consequently, spaces $\mathcal{U}$ and $\mathcal{V}$ must share common dimensionality,
i.e. $n=m$. By geometrical reasoning, the two volume elements are now related via
\begin{eqnarray}
|d\mathbf{v}|~=~\left|J(\mathbf{u})\right|\,|d\mathbf{u}|\,,
&\;\;\;\;&
J~\doteq~
\mathtt{det}\left[\frac{\partial\mathbf{h}}{\partial\mathbf{u}^T}\right]
\,,
\end{eqnarray}
where $J$ is known as the *Jacobian* of the transformation.

### Conservation of probability

In addition to the transformation geometry of the previous
[section](#Transformation-geometry "Section: Transformation geometry"), we now suppose
that spaces $\mathcal{U}$ and $\mathcal{V}$ have associated (but as yet unspecified) probability distributions denoted by $p:\mathcal{U}\mapsto\mathbb{R}^{+}$ and
$q:\mathcal{V}\mapsto\mathbb{R}^{+}$, respectively. 
We initially assume that $p$ and $q$ are *proper* distributions, such that they obey
$\int_\mathcal{U}p(\mathbf{u})\,|d\mathbf{u}|=1$ and
$\int_\mathcal{V}q(\mathbf{v})\,|d\mathbf{v}|=1$, respectively. 
It then follows that however we partition $\mathcal{U}$ and $\mathcal{V}$, the total probability mass for each must always be unity.
This is the key to the first invariance principle of conservation of probability mass.

Note, however, that we shall later relax this constraint (without justification), allowing $p$ and $q$
to be *improper* distributions that do not normalise to unity (in fact, their normalisation factors will be infinite).

Suppose we notionally partition $\mathcal{U}$ into a finite number $N$ of volume elements,
namely $(\delta U_1,\ldots,\delta U_N)$, such that $\mathcal{U}=\bigcup_{i=1}^{N}\delta U_i$.
Then the total probability mass must remain constant, viz:
\begin{eqnarray}
\int_\mathcal{U}p(\mathbf{u})\,|d\mathbf{u}| & = &
\sum_{i=1}^{N}\int_{\delta U_i}p(\mathbf{u})\,|d\mathbf{u}|~=~1\,.
\end{eqnarray}

Next, we apply transformation $\mathbf{h}$ to continuously map each volume element
$\delta U_i$ into its counterpart $\delta V_i$. Since we assumed that $\mathbf{h}$ is invertible, then the volume elements in $\mathcal{V}$ cannot overlap. Additionally, since $\mathcal{U}$ is mapped entirely into $\mathcal{V}$, there cannot be any space left over. Consequently, we must have partitioned $\mathcal{V}$ as $\mathcal{V}=\bigcup_{i=1}^{N}\delta V_i$, such that
\begin{eqnarray}
\int_\mathcal{V}q(\mathbf{v})\,|d\mathbf{v}| & = &
\sum_{i=1}^{N}\int_{\delta V_i}q(\mathbf{v})\,|d\mathbf{v}|~=~1\,.
\end{eqnarray}
It therefore follows that
\begin{eqnarray}
\sum_{i=1}^{N}\left(
\int_{\delta U_i}p(\mathbf{u})\,|d\mathbf{u}|-
\int_{\delta V_i}q(\mathbf{v})\,|d\mathbf{v}|
\right)
& = & 0\,.
\end{eqnarray}
Now, since this relation holds for arbitrary $N$, it must hold for all $N$, such that
\begin{eqnarray}
\int_{\delta U}p(\mathbf{u})\,|d\mathbf{u}| & = &
\int_{\delta V}q(\mathbf{v})\,|d\mathbf{v}|
\,.
\end{eqnarray}
In particular, in the limit as $N\rightarrow\infty$ we have $\delta U\rightarrow dU$
and $\delta V\rightarrow dV$, such that the relation
reduces to
\begin{eqnarray}
p(\mathbf{u})\,|d\mathbf{u}| & = & q(\mathbf{v})\,|d\mathbf{v}|
\,.
\end{eqnarray}

From the transformation geometry of the previous 
[section](#Transformation-geometry "Section: Transformation geometry"), 
we therefore obtain the standard probability conservation relation, viz:
\begin{eqnarray}
p(\mathbf{u}) & = & \left|J(\mathbf{u})\right|\,q(\mathbf{\mathbf{h}(\mathbf{u})})\,,
\end{eqnarray}
for all $\mathbf{u}\in\mathcal{U}$.

### Distributional invariance

The second invariance property is the key to the entire method of transformation groups. 
Consider 
two observers, X and Y, who both observe the same event, but who take measurements
$\mathbf{u}\in\mathcal{U}$ and $\mathbf{v}\in\mathcal{V}$ of the event, respectively.
We suppose that before the event both observers had exactly the same background knowledge.
The invariance is to suppose that after the event, both observers still have the same knowledge, since they observed the same event. Clearly, we are excluding imperfect measurements where observation was partially obscured for either observer.

As noted in the intoduction, 
Jaynes [[1]](#References "Reference: Prior probabilities and transformation groups") 
opined that if the 
knowledge of the two observers remains the same, then the observations are invariant to nature of the transformation, and furthermore observers X and Y must (for the sake of consistency) assign the same distribution to the event. Hence, Jaynes supposes that, identically, $q=p$.

Milne [[2]](#References "Reference: A note on scale invariance") 
disagreed, and demonstrated that although this invariance works for the linear scaling $y=\alpha x$, it fails to work
for the nonlinear scaling $y=\alpha x^\beta$. Hence, Milne proposed that the invariance instead holds up to a constant factor. 
As an aside, note that Milne used $f$ and $g$ in place of $p$ and $q$, respectively, and explicitly stated the constraint as $f=\gamma g$; however, his subsequent working out only makes sense if $g=\gamma f$ instead.

Indeed, as noted in the introduction, Jaynes himself saw the occasional need for $q=\gamma p$, as explicitly used in his solution 
[[3]](#References "Reference: The well-posed problem") 
to the Bertrand paradox 
(or supposed 'solution', since
Drory 
[[4]](#References "Reference: Failure and uses of Jaynes’ principle of transformation groups") 
counters Jaynes' claim to have a unique solution).

To consolidate these two viewpoints, we may informally suppose that invariance means here that the transformation does not alter the shape of the probability distribution, but only moves points along it, and possibly changes its constant of normalisation. Hence, we define
*distributional invariance* as
\begin{eqnarray}
q(\mathbf{v}) & = & \gamma p(\mathbf{v})\,,
\end{eqnarray}
for all $\mathbf{v}\in\mathcal{V}$, and for some constant $\gamma>0$.
Note that this now presupposes that $\mathcal{V}\subseteq\mathcal{U}$,
since $p:\mathcal{U}\mapsto\mathbb{R}^+$.

It now follows that if $p$ and $q$ are *proper* distributions, 
as assumed in the previous 
[section](#Conservation-of-probability "Section: Conservation of probability"),
then $\gamma$ is determined uniquely via conservation of probability as
\begin{eqnarray}
\gamma & = & 
\frac{\int_\mathcal{V}q(\mathbf{v})\,|d\mathbf{v}|}
{\int_\mathcal{V}p(\mathbf{v})\,|d\mathbf{v}|}
~=~\frac{1}{\int_\mathcal{V}p(\mathbf{u})\,|d\mathbf{u}|}\,.
\end{eqnarray}
Clearly, we must then have $\gamma=1$ if $\mathcal{V}=\mathcal{U}$, and
$\gamma>1$ if $\mathcal{V}\subset\mathcal{U}$.

However, if the distributions are *improper*, then $\gamma$ must be determined
from the above invariance relation, once the forms of $p$ and $q$ have been found.

### Unconditional probability invariance

The third invariance follows from the distributional invariance of the previous
[section](#Distributional-invariance "section: Distributional invariance"). 
Stated briefly, if the distribution is invariant to the transformation (up to a normalising constant), then it is invariant to the specific values of the transformation parameters (or at least a subset of the parameters, as we shall see in a later 
[example](#Nonlinear-rescaling-plus-relocation "Section: Nonlinear rescaling plus relocation")). 

Let the transformation $\mathbf{h}$ now be redefined explicitly as a function
$\mathbf{h}:\mathcal{U}\times\Psi\mapsto\mathcal{V}$
of parameters $\boldsymbol{\psi}\in\Psi$, such that 
$\mathbf{v}=\mathbf{h}(\mathbf{u};\boldsymbol{\psi})$, and likewise
$J(\mathbf{u};\boldsymbol{\psi})=\mathtt{det}(\partial\mathbf{h}/\partial\mathbf{u}^T)$.
Then the combination of 
[conservation of probability](#Conservation-of-probability "Section: Conservation of probability")
and 
[distributional invariance](#Distributional-invariance "Section: Distributional invariance")
requires that the (unconditional) probability distribution $p(\mathbf{u})$ 
obeys the unified invariance relation
\begin{eqnarray}
p(\mathbf{u}) & = & 
\gamma\,|J(\mathbf{u};\boldsymbol{\psi})|\,p(\mathbf{h}(\mathbf{u};\boldsymbol{\psi}))\,,
\end{eqnarray}
for every point $\mathbf{u}\in\mathcal{U}$, and *typically* (see the caveat further below)
for all parameter values $\psi\in\Psi$. We shall call this the *unconditional (probability) invariance* relation.

Since this relation is invariant to changes in the transformation parameter values $\boldsymbol{\psi}$, we may take derivatives with respect to 
$\boldsymbol{\psi}$, giving rise to the parameter invariance relation(s)
\begin{eqnarray}
\frac{\partial |J|}{\partial\boldsymbol{\psi}}(\mathbf{u};\boldsymbol{\psi})\, 
p(\mathbf{h}(\mathbf{u};\boldsymbol{\psi}))
+\left|J(\mathbf{u};\boldsymbol{\psi})\right|
\, \frac{\partial\mathbf{h}}
{\partial\boldsymbol{\psi}^T}(\mathbf{u};\boldsymbol{\psi})
\, 
\frac{\partial p}{\partial\mathbf{u}}
(\mathbf{h}(\mathbf{u};\boldsymbol{\psi}))& = & \mathbf{0}\,.
\end{eqnarray}
These invariances again hold for all $\mathbf{u}\in\mathcal{U}$, and typically for all
$\boldsymbol{\psi}\in\Psi$ (as per the caveat below).

Furthermore, if these derivatives are invariant to the values of the parameters, then the relations hold for all parameter values, including the special value $\boldsymbol{\psi}_0$ which leads to the identity mapping 
$\mathbf{h}(\mathbf{u};\boldsymbol{\psi}_0)=\mathbf{u}$.
Consequently, we may simplify the derivative invariances at 
$\boldsymbol{\psi}=\boldsymbol{\psi}_0$, and then solve the simplified equations for $p(\mathbf{u})$.

**Caveat**: Note that we stated above that the invariance relation in $\mathbf{u}$ is also typically satisfied for all transformation parameters 
$\boldsymbol{\psi}$. However, in some 
circumstances, as we shall see in the 
[example](#Nonlinear-rescaling-plus-relocation "Section: Nonlinear rescaling plus relocation") 
on nonlinear rescaling,
the invariance relation might be satisfied only for variation in some transformation parameters but not others. When this occurs, we have a *family* of invariance relations
dependent on (the constant values of) the non-invariant parameters. This includes the normalisation factor $\gamma$, which then more generally becomes 
$\gamma(\boldsymbol{\psi})$.
Consequently, the more general invariance relation is in fact
\begin{eqnarray}
p(\mathbf{u}) & = & 
\gamma(\boldsymbol{\psi})\,|J(\mathbf{u};\boldsymbol{\boldsymbol{\psi}})|\,
p(\mathbf{h}(\mathbf{u};\boldsymbol{\psi}))\,,
\end{eqnarray}
for all $\mathbf{u}\in\mathcal{U}$, and for all $\boldsymbol{\psi}\in\Psi$.

### Conditional probability invariance

So far, we have considered only the unconditional probability distributions $p(\mathbf{u})$ and $q(\mathbf{v})$. As Jaynes repeatedly stresses, all distributions are conditional in the sense that they depend (at the minimum) upon prior information about the problem to be solved (including, but not limited to, the respective domains $\mathcal{U}$ and
$\mathcal{V}$). Hence, by *unconditional* we mean not conditioned on any additional (i.e. exogneous) parameters.

This now leads us to consider conditional (i.e. parameterised) likelihood distributions of the form 
$p(\mathbf{u}\mid\boldsymbol{\theta})$ and $q(\mathbf{v}\mid\boldsymbol{\phi})$. 
In addition, we also have unconditional prior distributions of the form
$p(\boldsymbol{\theta})$ and $q(\boldsymbol{\phi})$.
The prior distributions are handled by the unconditional invariance relation of the previous
[section](#Unconditional-probability-invariance "Section: Unconditional probability invariance").
In particular, we consider the transformation 
$\mathbf{g}:\Theta\times\Psi\mapsto\Phi$ that maps distributional parameters $\boldsymbol{\theta}\in\Theta$ to 
$\boldsymbol{\phi}=\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})\in\Phi\subseteq\Theta$. 
The corresponding invariance relation is then
\begin{eqnarray}
p(\boldsymbol{\theta}) ~=~ \gamma_\mathbf{g}\,
\left|
J_\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})\right|
\,
p(\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))
\,,
&\;\;\;\;&
J_\mathbf{g}~\doteq~\mathtt{det}\left[\frac{\partial\mathbf{g}}{\partial\boldsymbol{\theta}^T}\right]
\,,
\end{eqnarray}
for all $\boldsymbol{\theta}\in\Theta$ and all $\psi\in\Psi$, with constant $\gamma_\mathbf{g}>0$.
Note for later use that the infinitesimal volumes $|d\boldsymbol{\theta}|$ and $|d\boldsymbol{\phi}|$ are related
by $|d\boldsymbol{\phi}|~=~|J_\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})|\,|d\boldsymbol{\theta}|$,
from the 
[transformation geometry](#Transformation-geometry "Section: Transformation geometry").

Next, we consider the joint distributions, namely
$p(\mathbf{u},\boldsymbol{\theta})=p(\mathbf{u}\mid\boldsymbol{\theta})\,p(\boldsymbol{\theta})$ and
$q(\mathbf{v},\boldsymbol{\phi})=q(\mathbf{v}\mid\boldsymbol{\phi})\,q(\boldsymbol{\phi})$, which in this form are also
to be considered unconditional, since there are no further exogenous parameters. 
To the existing transformation $\mathbf{g}$, we thus
add the adjusted transformation 
$\mathbf{h}:\mathcal{U}\times\Theta\times\Psi\mapsto\mathcal{V}$, which maps $\mathbf{u}\in\mathcal{U}$ to 
$\mathbf{v}=\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})\in\mathcal{V}\subseteq\mathcal{U}$.

From the
[transformation geometry](#Transformation-geometry "Section: Transformation geometry"), we deduce that the infinitesimal joint displacement $d(\mathbf{u},\boldsymbol{\theta})$ is related to the corresponding joint displacement 
$d(\mathbf{v},\boldsymbol{\phi})$ via
\begin{eqnarray}
\left[\begin{array}{c}
d\mathbf{v}
\\
d\boldsymbol{\phi}
\end{array}\right]
& = &
\left[\begin{array}{cc}
\frac{\partial\mathbf{h}}{\partial\mathbf{u}^T} &
\frac{\partial\mathbf{h}}{\partial\boldsymbol{\theta}^T}
\\
\mathbf{0} &
\frac{\partial\mathbf{g}}{\partial\boldsymbol{\theta}^T}
\end{array}\right]
\left[\begin{array}{c}
d\mathbf{u}
\\
d\boldsymbol{\theta}
\end{array}\right]\,.
\end{eqnarray}
The Jacobian of the joint transformation is therefore
\begin{eqnarray}
J_{\mathbf{h},\mathbf{g}} & \doteq & \mathtt{det}
\left[\begin{array}{cc}
\frac{\partial\mathbf{h}}{\partial\mathbf{u}} &
\frac{\partial\mathbf{h}}{\partial\boldsymbol{\theta}^T}
\\
\mathbf{0} &
\frac{\partial\mathbf{g}}{\partial\boldsymbol{\theta}^T}
\end{array}\right]
~=~\mathtt{det}\left[
\frac{\partial\mathbf{h}}{\partial\mathbf{u}^T}\right]\,
\mathtt{det}\left[
\frac{\partial\mathbf{g}}{\partial\boldsymbol{\theta}^T}\right]
\,,
\\\Rightarrow
J_{\mathbf{h},\mathbf{g}}(\mathbf{u},\boldsymbol{\theta};\boldsymbol{\psi})
& ~\equiv~ & 
J_\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})\,J_\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})\,.
\end{eqnarray}

Consequently, the infinitesimal joint volumes $|d(\mathbf{u},\boldsymbol{\theta})|$ and 
$|d(\mathbf{v},\boldsymbol{\phi})|$ are related via
\begin{eqnarray}
&
|d(\mathbf{v},\boldsymbol{\phi})| & ~=~ 
|J_{\mathbf{h},\mathbf{g}}(\mathbf{u},\boldsymbol{\theta};\boldsymbol{\psi})|\,|d(\mathbf{u},\boldsymbol{\theta})|
\\\Rightarrow~ &
|d\mathbf{v}|\,|d\boldsymbol{\phi}|
& ~=~ |J_\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})|\,|d\mathbf{u}|\,
|J_\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})|\,|d\boldsymbol{\theta}|\,,
\\\Rightarrow~ &
|d\mathbf{v}| & ~=~  |J_\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})|\,|d\mathbf{u}|\,,
\end{eqnarray}
since we noted earlier that $|d\boldsymbol{\phi}|=|J_\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})|\,|d\boldsymbol{\theta}|$ for
the prior distributions.
Hence, the relation above holds for the conditional distributions, such that the joint invariance must factor into a prior invariance and a conditional invariance.

To see this in more detail, note that the invariance relation for the joint distribution(s) is
\begin{eqnarray}
& p(\mathbf{u},\boldsymbol{\theta}) & ~=~ \gamma_{\mathbf{h},\mathbf{g}}\,
|J_{\mathbf{h},\mathbf{g}}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})|\,
p(\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi}),\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))
\\\Rightarrow~ &
p(\mathbf{u}\mid\boldsymbol{\theta})\,p(\boldsymbol{\theta}) & ~=~ \gamma_{\mathbf{h},\mathbf{g}}\,
|J_\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})|\,
p(\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})\mid\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))\,
|J_\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})|\,
p(\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))\,,
\\\Rightarrow~ &
\gamma_\mathbf{g}\,p(\mathbf{u}\mid\boldsymbol{\theta}) & ~=~
\gamma_{\mathbf{h},\mathbf{g}}\,
|J_\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})|\,
p(\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})\mid\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))\,,
\end{eqnarray}
where the last line follows directly from the prior invariance relation
given earlier, namely 
$p(\boldsymbol{\theta})=\gamma_\mathbf{g}|J_\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})|\,p(\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))$.
Hence, letting $\gamma_{\mathbf{h},\mathbf{g}}=\gamma_\mathbf{h}\gamma_\mathbf{g}$, 
we finally obtain the *conditional (probability) invariance* relation
\begin{eqnarray}
p(\mathbf{u}\mid\boldsymbol{\theta}) & = & \gamma_\mathbf{h}\,
|J_\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})|\,
p(\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})\mid\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))\,,
\end{eqnarray}
which holds for all $\mathbf{u}\in\mathcal{U}$, all $\boldsymbol{\theta}\in\Theta$, and 
*typically* for all $\boldsymbol{\psi}\in\Psi$. 

However, from the caveat given in
the previous 
[section](#Unconditional-probability-invariance "Section: Unconditional probability invariance"),
if invariance does **not** in fact hold for all subsets of the transformation parameter $\boldsymbol{\psi}$, then we require the more general invariance relation
\begin{eqnarray}
p(\mathbf{u}\mid\boldsymbol{\theta}) & = & \gamma_\mathbf{h}(\boldsymbol{\psi})\,
|J_\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})|\,
p(\mathbf{h}(\mathbf{u};\boldsymbol{\theta},\boldsymbol{\psi})\mid\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}))\,,
\end{eqnarray}
which **does** now hold for all $\boldsymbol{\psi}\in\Psi$.

### Summary of invariances

The invariance properties derived in the previous sections appear highly complex, due to the need to handle various technicalities. However, it is possible to restate them in a simpler form closer to their original intent. 
For instance, to emphasise the fact that invariant transformations are 
[required](#Distributional-invariance "Section: Distributional invariance")
to be self-mapping, viz $\Phi\subseteq\Theta$ and 
$\mathcal{V}\subseteq\mathcal{U}$, we replace 
$\Phi$ by $\Theta$ and both $\mathcal{U}$ and $\mathcal{V}$ by
$\mathcal{X}$. Similarly, we replace 
$\boldsymbol{\phi},\mathbf{u},\mathbf{v}$ by
$\boldsymbol{\theta}',\mathbf{x},\mathbf{x}'$, respectively, where the prime
denotes the image of the corresponding unprimed variable.
The key invariances then take the form:

1. The prior distribution $p(\boldsymbol{\theta})$ 
is invariant to the transformation 
$\mathbf{g}:\Theta\times\Psi\mapsto\Theta$, with image
$\boldsymbol{\theta}'\doteq\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi})$,
provided that
\begin{eqnarray}
p(\boldsymbol{\theta}) & ~=~ & 
\gamma_g\,\left|\,\mathtt{det}
\frac{\partial\mathbf{g}}{\partial\boldsymbol{\theta}^T}\,
\right|\,
p(\boldsymbol{\theta}')
\end{eqnarray}
holds for all $\boldsymbol{\theta}\in\Theta$ and $\boldsymbol{\psi}\in\Psi$,
where the distribution function $p(\cdot)$ on both sides is identical.
Note that the proportionality $\gamma_g$ may also be a function of $\boldsymbol{\psi}$,
as necessary.

2. The likelihood distribution $p(\mathbf{x}\mid\boldsymbol{\theta})$ 
is invariant to the transformation
$\mathbf{h}:\mathcal{X}\times\Theta\times\Psi\mapsto\mathcal{X}$,
with image
$\mathbf{x}'=\mathbf{h}(\mathbf{x};\boldsymbol{\theta},\boldsymbol{\psi})$,
provided that
\begin{eqnarray}
p(\mathbf{x}\mid\boldsymbol{\theta}) & ~=~ & 
\gamma_h\,\left|\,\mathtt{det}
\frac{\partial\mathbf{h}}{\partial\mathbf{x}^T}\,
\right|\,
p(\mathbf{x}'\mid\boldsymbol{\theta}')
\end{eqnarray}
holds for all $\mathbf{x}\in\mathcal{X}$, 
$\boldsymbol{\theta}\in\Theta$ and $\boldsymbol{\psi}\in\Psi$,
where the distribution function $p(\cdot\mid\cdot)$ on both sides is identical.
Note that the proportionality $\gamma_h$ may also be a function of 
$\boldsymbol{\psi}$,
as necessary.

In order to determine the functional forms of $p(\cdot)$ and $p(\cdot\mid\cdot)$, we proceed as follows:

1. Take derivatives of the invariance relations with respect to 
$\boldsymbol{\psi}$, and evaluate the
results at the point $\boldsymbol{\psi}_0$ of identity where
$\mathbf{g}(\boldsymbol{\theta};\boldsymbol{\psi}_0)=\boldsymbol{\theta}$
for all $\boldsymbol{\theta}\in\Theta$, and 
$\mathbf{h}(\mathbf{x};\boldsymbol{\theta},\boldsymbol{\psi}_0)=\mathbf{x}$
for all $\mathbf{x}\in\mathcal{X}$ and $\boldsymbol{\theta}\in\Theta$.

2. Solve the resulting differential equations for $p(\boldsymbol{\theta})$
and $p(\mathbf{x}\mid\boldsymbol{\theta})$.

3. Substitute these functions back into the original invariance relations
and solve for $\gamma_g$ and $\gamma_h$.

Note that if $\gamma_g$ and/or $\gamma_h$ are required to also be functions of
$\boldsymbol{\psi}$, then step 2 will provide additional constraints
on their functional forms.

## Examples of Invariance

### Nonlinear rescaling plus relocation

Consider a scaling parameter $\sigma\in\mathbb{R}^{+}$ 
and a location parameter $\mu\in\mathbb{R}$, such that the
prior distriubtion $p(\sigma,\mu)$ is invariant both to further rescaling and to translation.

Jaynes [[1]](#References "Reference: Prior probabilities and transformation groups")
solved this problem for simple translation $\mu'=\mu+\nu$ and linear rescaling
$\sigma'=\alpha\sigma$. However, 
Milne [[2]](#References "Reference: A note on scale invariance")
solved the rescaling invariance problem (without translation) for nonlinear
scaling $\sigma'=\alpha\sigma^\beta$. Here we combine both approaches.

Let the prior distribution $p(\mu)$ be invariant to the transformation
\begin{eqnarray}
f(\mu;\nu) & = & \mu+\nu\,,
\end{eqnarray}
which has Jacobian
\begin{eqnarray}
J_f(\mu;\nu) & = & \frac{\partial f}{\partial\mu}~=~1\,.
\end{eqnarray}
Note that the transformation becomes the identity for $\nu=\nu_0=0$.

The [invariance relation](#Summary-of-invariances "Section: Summary of invariances")
is thus
\begin{eqnarray}
p(\mu) & = & \gamma_f\,p(\mu+\nu)\,.
\end{eqnarray}
Taking the derivative with respect to the transformation parameter $\nu$, and setting
$\nu=\nu_0$, gives
\begin{eqnarray}
0~=~\gamma_f\,p'(\mu) & \Rightarrow & p(\mu)~=~k_f\,,
\end{eqnarray}
for abitrary constant $k_f>0$. Note that this is an improper prior distribution.
Substitution back into the invariance relation then gives $\gamma_f=1$.

Next, suppose that the prior distribution $p(\sigma)$ is invariant to the transformation
\begin{eqnarray}
g(\sigma;\alpha,\beta) & = & \alpha\sigma^\beta\,,
\end{eqnarray}
with $\alpha>0$ and $\beta\ne 0$.
Note that the transformation becomes the identity 
at the point $(\alpha_0,\beta_0)=(1,1)$.
The Jacobian of the transformation is then
\begin{eqnarray}
J_g(\mu;\alpha,\beta) & = & \frac{\partial g}{\partial\sigma}
~=~\alpha\beta\sigma^{\beta-1}\,,
\end{eqnarray}
such that the
[invariance relation](#Summary-of-invariances "Section: Summary of invariances") is given by
\begin{eqnarray}
p(\sigma) & = & \gamma_g\,\alpha|\beta|\sigma^{\beta-1}\,p(\alpha\sigma^\beta)\,.
\end{eqnarray}
Taking derivatives with respect to the transformation parameters $\alpha$ and $\beta$
therefore give
\begin{eqnarray}
0 & = & \gamma_g\, |\beta|\sigma^{\beta-1}\left[
p(\alpha\sigma^\beta)
 +\alpha\,p'(\alpha\sigma^\beta)\,\sigma^\beta
\right]\,,
\\
0 & = & \gamma_g\,\alpha\,\sigma^{\beta-1}\left[
 \mathtt{sign}(\beta)\,p(\alpha\sigma^\beta)
 +|\beta|\,\ln\sigma\,p(\alpha\sigma^\beta)
 +|\beta|\,p'(\alpha\sigma^\beta)\,\alpha\sigma^\beta\,\ln\sigma
\right]\,,
\end{eqnarray}
respectively.
Next, by substituting $(\alpha,\beta)=(\alpha_0,\beta_0)$, and simplifying the results, we obtain
\begin{eqnarray}
p(\sigma)+\sigma\,p'(\sigma) & = & 0\,,
\\
\left[1+\ln\sigma\right]\,p(\sigma)+\sigma\ln\sigma\,p'(\sigma) & = & 0\,.
\end{eqnarray}

It turns out that we cannot (usefully) satisfy both of these equations simultaneously.
The former equation (due to variation in $\alpha$) has the solution
\begin{eqnarray}
p(\sigma) & = & \frac{k_g}{\sigma}\,,
\end{eqnarray}
for arbitrary constant $k_g>0$. However, the latter equation (due to variation in $\beta$) then reduces to $k_g=0$.
Consequently, we see that the nonlinear scaling **is** invariant to $\alpha>0$, but is **not** invariant to $\beta$, which must therefore be held constant to a value fixed in advance. In fact, from the invariance relation above, we have
\begin{eqnarray}
& p(\sigma) & ~=~ \gamma_g\,\alpha|\beta|\sigma^{\beta-1}\,p(\alpha\sigma^\beta)\,,
\\
\Rightarrow & \frac{k_g}{\sigma} & ~=~
 \gamma_g\,\alpha|\beta|\sigma^{\beta-1}\frac{k_g}{\alpha\sigma^\beta}\,,
\\
\Rightarrow & \gamma_g & ~=~ \frac{1}{|\beta|}\,.
\end{eqnarray}
This dependence of the renormalisation factor $\gamma_g$ on $\beta$ means that for each fixed value of $\beta$ we obtain a family of transformations that are distributionally invariant to the scaling factor $\alpha$.

In fact, we actually require the 
[general form](#Unconditional-probability-invariance "Section: Unconditional probability invariance")
of the invariance relation, 
obtained by explicitly making $\gamma_g$ a function of $(\alpha,\beta)$.
In this case, the relation and its derivatives become
\begin{eqnarray}
& p(\sigma) & ~=~ \gamma_g(\alpha,\beta)\,\alpha|\beta|\sigma^{\beta-1}\,p(\alpha\sigma^\beta)\,,
\\
\Rightarrow & 0 & ~=~  \left[
\alpha\frac{\partial\gamma_g}{\partial\alpha}+\gamma_g
\right]\,p(\sigma')+\gamma_g\alpha\sigma^\beta\,p'(\sigma')\,,
\\
& 0 & ~=~ \frac{\partial\gamma_g}{\partial\beta}|\beta|\,p(\sigma')
\\\nonumber&&
{}+
\gamma_g\left[
 \mathtt{sign}(\beta)\,p(\sigma')
 +|\beta|\,\ln\sigma\,p(\sigma')
 +|\beta|\,p'(\sigma')\,\alpha\sigma^\beta\,\ln\sigma
\right]\,,
\\
\Rightarrow & 0 & ~=~ \frac{\partial\gamma_g}{\partial\beta}|\beta|
+\gamma_g\,\mathtt{sign}(\beta)-
\alpha\frac{\partial\gamma_g}{\partial\alpha}\,|\beta|\ln\sigma\,.
\end{eqnarray}
However, since $\gamma_g(\alpha,\beta)$ is not a function of $\sigma$, then 
\begin{eqnarray}
\frac{\partial\gamma_g}{\partial\alpha} & = & 0 ~~\Rightarrow~~
\frac{\partial\gamma_g}{\partial\beta}
~=~-\gamma_g\frac{\mathtt{sign}(\beta)}{|\beta|}
~~\Rightarrow~~\gamma_g~=~\frac{1}{|\beta|}
\,.
\end{eqnarray}

We see that taking $\beta=1$ results in the linear scaling of 
Jaynes [[1]](#References "Reference: Prior probabilities and transformation groups"),
for which $\gamma_g=1$. However, $\gamma_g\ne 1$ in general for $|\beta|\ne 1$, so the correct form of the invariance prior is
\begin{eqnarray}
p(\sigma) & ~\propto~ & \frac{1}{|\beta|\,\sigma}\,.
\end{eqnarray}
Despite the fact that this is an *improper* distribution that cannot actually be normalised, we maintain that
[conservation of probability](#Conservation-of-probability "Section: Conservation of probability") continues to hold.

Before we move on to the likelihood function, we might ask the question: What would the prior be if we parameterised in terms of $\sigma^2$ instead of 
$\sigma$? To answer this in general, consider $\sigma^n$. Then we see that,
under the same transformation $g$, we have
\begin{eqnarray}
\sigma~\rightarrow~\alpha\,\sigma^\beta & ~~~\Rightarrow~~~ &
\sigma^n~\rightarrow~(\alpha\,\sigma^\beta)^n~=~\alpha^n\,(\sigma^n)^\beta\,.
\end{eqnarray}
Hence, since $\alpha$ is a just constant parameter, we may repeat the above derivation in terms of $\alpha^n$ and $\sigma^n$ instead of $\alpha$ and 
$\sigma$. Consequently, we obtain
\begin{eqnarray}
p(\sigma^n) & ~\propto~ & \frac{1}{|\beta|\,\sigma^n}\,.
\end{eqnarray}
This demonstrates that the prior really is invariant to scaling.

Finally, we now turn to the subsequent problem posed by
Jaynes [[1]](#References "Reference: Prior probabilities and transformation groups"),
namely that of invariance of the likelihood distribution 
$p(x\mid\sigma,\mu)$ to
rescaling and relocation, specifically of the form $x'=(\sigma'/\sigma)(x-\mu)+\mu'$.
Allowing for nonlinear rescaling, we therefore suppose that $p(x\mid\sigma,\mu)$
is invariant to the transformation
\begin{eqnarray}
h(x;\sigma,\mu;\alpha,\beta,\nu) & = & \alpha\sigma^{\beta-1}\,x
-(\alpha\sigma^{\beta-1}-1)\,\mu+\nu\,.
\end{eqnarray}
Note that this transformation becomes the identity for 
$(\alpha,\beta,\nu)=(\alpha_0,\beta_0,\nu_0)$. Also note that its Jacobian is
\begin{eqnarray}
J_h(x;\sigma,\mu;\alpha,\beta,\nu) & = & \frac{\partial h}{\partial x}
~=~\alpha\sigma^{\beta-1}\,.
\end{eqnarray}

Hence, the [invariance relation](#Summary-of-invariances "Section: Summary of invariances") is given by
\begin{eqnarray}
p(x\mid\sigma,\mu) & ~=~ & \gamma_h\,|J_h|\,p(x'\mid\sigma',\mu')
\\
& ~=~ & \gamma_h\,\alpha\sigma^{\beta-1}\,
p(\alpha\sigma^{\beta-1}\,(x-\mu)+\mu+\nu
\mid\alpha\sigma^\beta,\mu+\nu)\,.
\end{eqnarray}
We observed above that the transformation $g$ for $\sigma$ is not invariant to variation
in the transformation parameter $\beta$. Hence, taking only derivatives with respect to the other transformation parameters, $\alpha$ and $\nu$, and simplifying, we obtain
\begin{eqnarray}
0 & = & 
  p(x'\mid\sigma',\mu')
 +\alpha\,\frac{\partial p}{\partial x}(x'\mid\sigma',\mu')
  \,\sigma^{\beta-1}(x-\mu)
 +\alpha\,\frac{\partial p}{\partial\sigma}(x'\mid\sigma',\mu')
  \,\sigma^\beta\,,
\\
0 & = & 
  \frac{\partial p}{\partial x}(x'\mid\sigma',\mu')
 +\frac{\partial p}{\partial\mu}(x'\mid\sigma',\mu')\,,
\end{eqnarray}
respectively. Thus, at $(\alpha_0,\beta_0,\nu_0)$ we have
\begin{eqnarray}
p+(x-\mu)\frac{\partial p}{\partial x}+\sigma\frac{\partial p}{\partial\sigma} & = & 0\,,
\\
\frac{\partial p}{\partial x}+\frac{\partial p}{\partial\mu} & = & 0\,.
\end{eqnarray}
The latter equation indicates that $p(x\mid\sigma,\mu)$ is a function of $x-\mu$, and
the former equation is then satisfied in general for
\begin{eqnarray}
p(x\mid\sigma,\mu) & = & \frac{k_h}{\sigma}\varphi\left(\frac{x-\mu}{\sigma}\right)\,,
\end{eqnarray}
where $\varphi(\cdot)$ is known as a *radial basis function*. Observe that this
transformation invariance therefore holds for the Gaussian distribution with basis
function
$\varphi(z)=e^{-\frac{1}{2}z^2}$.

Finally, substituting this back into the conditional invariance relation, we obtain
\begin{eqnarray}
\frac{k_h}{\sigma}\varphi\left(\frac{x-\mu}{\sigma}\right)
~=~
\gamma_h\,\alpha\sigma^{\beta-1}\,\frac{k_h}{\alpha\sigma^\beta}
\varphi\left(\frac{x-\mu}{\sigma}\right)
& \;\;\Rightarrow\;\; & \gamma_h~=~1\,.
\end{eqnarray}

Lastly, just for the sake of consistency, we check the derivative of the invariance relation with respect to $\beta$, namely
\begin{eqnarray}
0 & = & 
  p(x'\mid\sigma',\mu')
 +\frac{\partial p}{\partial x}(x'\mid\sigma',\mu')\,
  \alpha\sigma^{\beta-1}(x-\mu)
+\frac{\partial p}{\partial\sigma}(x'\mid\sigma',\mu')\,
  \alpha\sigma^\beta\,.
\end{eqnarray}
Evaluating this at $(\alpha_0,\beta_0,\nu_0)$ gives
\begin{eqnarray}
  p+(x-\mu)\,\frac{\partial p}{\partial x}
 +\sigma\frac{\partial p}{\partial\sigma} & = & 0
\,.
\end{eqnarray}
We recognise this as being exactly the invariance relation for $\alpha$, and hence is entirely consistent but superfluous. In fact, this consistency 
means that $p(x\mid\sigma,\mu)$ is also invariant to changes in $\beta$, and
is thus the reason that $\gamma_h$ is not a function of $\beta$,
unlike $\gamma_g$.

### Von Kries paradox

We now turn to the resolution of the Von Kries paradox, as also discussed (briefly) by 
Milne [[2]](#References "Reference: A note on scale invariance").

Suppose we know that the specific density $\rho$ (my notation) of a fluid is restricted to the range $\rho\in[a,b]$ for some $0<a<b<\infty$. If we know nothing else, then the principle of insufficient reason suggests all values are equally likely, and so we might take the uniform prior $f(\rho)=\frac{1}{b-a}$.

However, the specific volume $\nu$ (again, my notation) is inversely related to the specific density via $\nu=\frac{1}{\rho}$, giving rise to the Jacobian
$J(\rho)=\frac{d\nu}{d\rho}=-\frac{1}{\rho^2}$. The corresponding prior $g(\nu)$ therefore satisfies the conservation of probability relation
\begin{eqnarray}
f(\rho)~=~|J(\rho)|\,g(\nu) & \Rightarrow & 
g(\nu)~=~\frac{1}{\nu^2}f\left(\frac{1}{\nu}\right)
~=~\frac{1}{(b-a)\nu^2}\,.
\end{eqnarray}
The paradox is therefore that the specific volume is deterministically known from the specific density, but uniform ignorance of the specific density does not translate to uniform ignorance of the specific volume.  In other words, we seem to have acquired some knowledge of the specific volume for free.

The resolution of the paradox is that we must already have added this knowledge by the *unwarranted* assumption of a uniform prior for the specific density.
Instead, we take ignorance to mean 
[distributional invariance](#Distributional-invariance "Section: Distributional invariance"). 
From the 
[previous](#Nonlinear-rescaling-plus-relocation "Section: Nonlinear rescaling plus relocation") section, 
we see that 
$\nu=\frac{1}{\rho}$ is a specific form of the nonlinear transformation
$\nu=\alpha\rho^\beta$ for $\alpha=1$ and $\beta=-1$. 
Consequently, we already know that $\gamma=\frac{1}{|\beta|}=1$, such that the two prior
distributions are given by
\begin{eqnarray}
f(\rho)~=~\frac{k}{\rho}\,, &\;\;\;\;& 
g(\nu)~=~\frac{1}{\nu^2}f\left(\frac{1}{\nu}\right)~=~\frac{k}{\nu}\,,
\end{eqnarray}
respectively. Since we assumed a finite domain $\rho\in[a,b]$, we can easily obtain
the normalisation constant as $k=\frac{1}{\ln b-\ln a}$.
We observe that now ignorance of the specific density gives rise to the same form as ignorance of the specific volume, which resolves the paradox.

### Linear regression coefficients

Let us consider (briefly) the problem of fitting the straight-line model 
$y=\alpha+\beta x$, for $x,y\in\mathbb{R}$. Typically, in practice, we observe pairs of values $(x, y)$, and from these data $D$ we wish to infer the posterior distribution $p(\alpha,\beta\mid D)$.
However, here we are not concerned with the data but with the prior distributions, $p(\alpha)$ and $p(\beta)$, of the regression coefficients.

Firstly, we observe that $\alpha\in\mathbb{R}$ uniquely specifies the intersection of the model line with the $y$-axis, and hence $\alpha$ is a location parameter.
Consequently, if we know nothing else about $\alpha$, then
we might suppose that the prior distribution, $p(\alpha)$, is invariant to the translation 
$\alpha'=\alpha+\nu$. The 
[invariance relation](#Summary-of-invariances "Section: Summary of invariances") is thus
\begin{eqnarray}
p(\alpha) & = & \gamma_1\,p(\alpha+\nu)\,.
\end{eqnarray}

As usual, we take the derivative with respect to the transformation parameter $\nu$, and then evaluate the result at the point $\nu_0=0$, for which the transformation becomes an identity. This gives
\begin{eqnarray}
0~=~\gamma_1\,p'(\alpha) & ~~~\Rightarrow~~~ & p(\alpha)~=~k_1\,,\;\;\gamma_1~=~1\,.
\end{eqnarray}
Thus, $\alpha$ has an improper, uniform invariance prior.

Next, we observe that the slope $\beta\in\mathbb{R}$ is the tangent of the angle
$\boldsymbol{\theta}\in(-\frac{\pi}{2},\frac{\pi}{2})$ that the model line is rotated
counter-clockwise from the $x$-axis.
Some trigonometry gives us, for example:
\begin{eqnarray}
\beta & = & \tan\boldsymbol{\theta} ~=~ \frac{\sin\boldsymbol{\theta}}{\cos\boldsymbol{\theta}}
~=~\frac{\mathtt{sign}(\boldsymbol{\theta})\,\sqrt{1-\cos^2\boldsymbol{\theta}}}{\cos\boldsymbol{\theta}}
\\
\Rightarrow \cos\boldsymbol{\theta} & = & \frac{1}{\sqrt{1+\beta^2}}\,.
\end{eqnarray}
Next, we see that
\begin{eqnarray}
\frac{d\tan\boldsymbol{\theta}}{d\boldsymbol{\theta}} & = & 
\frac{d}{d\boldsymbol{\theta}}\left(\frac{\sin\boldsymbol{\theta}}{\cos\boldsymbol{\theta}}\right)
~=~
\frac{\cos^2\boldsymbol{\theta}+\sin^2\boldsymbol{\theta}}{\cos^2\boldsymbol{\theta}}
\\& = &
1+\tan^2\boldsymbol{\theta}
~=~\frac{1}{\cos^2\boldsymbol{\theta}}~=~\sec^2\boldsymbol{\theta}\,,
\\
\Rightarrow \frac{d\beta}{d\boldsymbol{\theta}} & = & 1+\beta^2\,.
\end{eqnarray}

If we are ignorant about any further properties of $\beta$, then we may suppose that the
prior distribution, $p(\beta)$, is invariant to the rotational transformation
$\beta'=\tan(\boldsymbol{\theta}+\psi)$. The
[invariance relation](#Summary-of-invariances "Section: Summary of invariances") is then
\begin{eqnarray}
p(\beta)\,\left|\frac{d\beta}{d\boldsymbol{\theta}}\right| & = &
\gamma_2\,p(\beta')\,\left|\frac{\partial\beta'}{\partial\boldsymbol{\theta}}\right|\,,
\end{eqnarray}
where
\begin{eqnarray}
\frac{\partial\beta'}{\partial\boldsymbol{\theta}} & = &
\frac{\partial\beta'}{\partial\psi}~=~1+\beta'^2\,.
\end{eqnarray}
Hence, the completed relation is
\begin{eqnarray}
(1+\beta^2)\,p(\beta) & = &
\gamma_2\,(1+\beta'^2)\,p(\beta')\,.
\end{eqnarray}

Once again, we take the derivative with respect to the transformation parameter $\psi$. This gives
\begin{eqnarray}
0 & = & \gamma_2\,\left[
2\beta'(1+\beta'^2)\,p(\beta')+(1+\beta'^2)^2\,p'(\beta')
\right]\,.
\end{eqnarray}
Next, we evaluate the derivative at the point $\psi_0=0$ at which the transformation becomes the identity. With simplification, this gives
\begin{eqnarray}
& 0 & ~=~2\beta\,p(\beta)+(1+\beta^2)\,p'(\beta)\,,
\\
\Rightarrow & 
\frac{p'(\beta)}{p(\beta)} & ~=~ -\frac{2\beta}{1+\beta^2}\,,
\\
\Rightarrow &
\ln p(\beta) & ~=~ \ln k_2-\ln(1+\beta^2)\,,
\\
\Rightarrow &
p(\beta) & ~=~ \frac{k_2}{1+\beta^2}\,.
\end{eqnarray}
Substitution back into the invariance relation then gives $\gamma_2=1$.
Furthermore, we recognise that $p(\beta)$ is just the 
[Cauchy](https://en.wikipedia.org/wiki/Cauchy_distribution "Wikipedia: Cauchy distribution") distribution, which properly normalises with $k_2=\frac{1}{\pi}$.

### Time-scale invariance

Consider the conditional distribution $p(t\mid\lambda)$ for a temporal variate
$t\in[0,\infty)$ and a frequency parameter $\lambda\in(0,\infty)$. The proper scale of time is thus given by $\tau\doteq\lambda^{-1}$.
Next, as per Jaynes' suggestion, consider two observers, X and Y, whose watches run at different rates. Hence, observer X has time-scale $\tau$, but observer Y has time-scale
$\tau'=\alpha\tau$, for $\alpha>0$. Assuming the prior $p(\tau)$ is invariant to this transformation, we obtain, from the examples on 
[rescaling](#Nonlinear-rescaling-plus-relocation "Section: Nonlinear rescaling plus relocation")
and the [Von Kries paradox](#Von-Kries-paradox "Section: Von Kries paradox"),
that
\begin{eqnarray}
p(\tau)~=~\frac{k}{\tau}\,, &\;\;\;\;& p(\lambda)~=~\frac{k}{\lambda}\,.
\end{eqnarray}
Similarly, we see that if observer X measures time as $t$, then observer Y measures times as $t'=\frac{\tau'}{\tau} t=\alpha t$. Assuming that $p(t\mid\lambda)$ is invariant to this transformation, 
and defining $\lambda'\doteq\tau'^{-1}=\lambda/\alpha$
we obtain the
[invariance relation](#Summary-of-invariances "Section: Summary of invariances")
\begin{eqnarray}
p(t\mid\lambda) & = & \gamma\alpha\,p\left(\alpha t\;\left|\;\frac{\lambda}{\alpha}\right.\right)\,,
\end{eqnarray}
for some $\gamma>0$.

Hence, taking the derivative with respect to the transformation parameter $\alpha$
gives
\begin{eqnarray}
0 & = &  \gamma\left[
p(t'\mid\lambda')
+\alpha\,t\frac{\partial p}{\partial t}(t'\mid\lambda')
-\frac{\lambda}{\alpha}\frac{\partial p}{\partial\lambda}(t'\mid\lambda')
\right]\,.
\end{eqnarray}
Lastly, we evaluate this at $\alpha_0=1$, at which point the transformation is the identity, and simplify the result, giving
\begin{eqnarray}
p(t\mid\lambda)
+t\frac{\partial p}{\partial t}(t\mid\lambda)
-\lambda\frac{\partial p}{\partial\lambda}(t\mid\lambda)
& = & 0\,.
\end{eqnarray}
This has the solution
\begin{eqnarray}
p(t\mid\lambda) & = & \lambda e^{-\lambda t}\,,
\end{eqnarray}
which we recognise as the [exponential](https://en.wikipedia.org/wiki/Exponential_distribution "Wikipedia: Exponential distribution") distribution.
Hence, the exponential distribution is invariant to the time scale,
and the rate parameter $\lambda$ has an improper, inverse prior.

## Minimum Information Priors

Apart from the transformation group invariance discussed in the 
[previous](#Transformation-Group-Invariance "Section: Transformation Group Invariance") section,
there are other methodologies for choosing prior distributions on the grounds of *non-informativeness*. One such method, as discussed by Zellner [[5]](#References "Reference: An introduction to Bayesian inference in econometrics"),
is that of choosing the prior to minimise a formal notion of entropic information. 

### Minimum information principle

To set the scene, consider the likelihood distribution $p(\mathbf{x}\mid\boldsymbol{\theta})$,
for $\mathbf{x}\in\mathcal{X}$ and $\boldsymbol{\theta}\in\Theta$.
The differential entropy of this distribution is defined as
\begin{eqnarray}
H(\mathcal{X}\mid\boldsymbol{\theta}) & ~\doteq~ &
-\int_\mathcal{X} p(\mathbf{x}\mid\boldsymbol{\theta})\,\log p(\mathbf{x}\mid\boldsymbol{\theta})\,|d\mathbf{x}|\,,
\end{eqnarray}
which is conditional on some arbitrary but fixed value of the distributional parameters
$\boldsymbol{\theta}$. Hence, for some prior distribution $p(\boldsymbol{\theta})$, the average entropy is just the conditional entropy, defined as
\begin{eqnarray}
H(\mathcal{X}\mid\Theta) & ~\doteq~ &
\int_\Theta H(\mathcal{X}\mid\boldsymbol{\theta})\,p(\boldsymbol{\theta})\,|d\boldsymbol{\theta}|\,.
\end{eqnarray}
The prior distribution $p(\boldsymbol{\theta})$ itself has the entropy
\begin{eqnarray}
H(\Theta) & ~\doteq~ &
-\int_\Theta p(\boldsymbol{\theta})\,\log p(\boldsymbol{\theta})\,|d\boldsymbol{\theta}|\,,
\end{eqnarray}
such that the joint entropy is given by
\begin{eqnarray}
H(\mathcal{X},\Theta) & ~\doteq~ & 
-\int_\Theta\int_\mathcal{X} p(\mathbf{x},\boldsymbol{\theta})\,\log
p(\mathbf{x},\boldsymbol{\theta})\,|d\mathbf{x}|\,|d\boldsymbol{\theta}|
\\& ~=~ &
-\int_\Theta\int_\mathcal{X} p(\mathbf{x}\mid\boldsymbol{\theta})\,p(\boldsymbol{\theta})\,\log\left[
p(\mathbf{x}\mid\boldsymbol{\theta})\,p(\boldsymbol{\theta})\right]\,|d\mathbf{x}|\,|d\boldsymbol{\theta}|
\\& ~=~ &
-\int_\Theta p(\boldsymbol{\theta})\,\int_\mathcal{X} p(\mathbf{x}\mid\boldsymbol{\theta})\,\log p(\mathbf{x}\mid\boldsymbol{\theta})\,|d\mathbf{x}|\,|d\boldsymbol{\theta}|
\\&&
{}-\int_\Theta p(\boldsymbol{\theta})\,\log p(\boldsymbol{\theta})\,\int_\mathcal{X}
p(\mathbf{x}\mid\boldsymbol{\theta})\,|d\mathbf{x}|\,|d\boldsymbol{\theta}|
\\& = &
H(\mathcal{X}\mid\Theta)+H(\Theta)\,.
\end{eqnarray}

The maximum entropy principle now asserts that the distribution with the greatest entropy embodies the least prior information, such that a gain in information corresponds to a loss in entropy. In fact, Zellner [[5]](#References "Reference: An introduction to Bayesian inference in econometrics") would appear to define information as the negative of entropy.

We therefore consider the change in entropy specified via
\begin{eqnarray}
G(\mathcal{X},\Theta) & ~\doteq~ & H(\mathcal{X}\mid\Theta)-H(\Theta)\,.
\end{eqnarray}
Observe that as $p(\boldsymbol{\theta})$ becomes more informative, then
$H(\Theta)$ decreases and, roughly speaking, $G(\mathcal{X},\Theta)$ increases.
Thus, $G(\mathcal{X},\Theta)$ apparently measures the additional information contained in the prior distribution $p(\boldsymbol{\theta})$ that is not also present in the sampling distribution $p(\mathbf{x}\mid\boldsymbol{\theta})$. Consequently, minimising $G$ should correspond to the choosing the least informative prior $p(\boldsymbol{\theta})$, relative to
a fixed distribution $p(\mathbf{x}\mid\boldsymbol{\theta})$. 
This is the *minimum information principle*.

Subject to the constraint that 
\begin{eqnarray}
\int_\Theta p(\boldsymbol{\theta})\,|d\boldsymbol{\theta}| & ~=~ & 1\,,
\end{eqnarray}
we therefore introduce the Lagrangian multiplier $\lambda$, and minimise the functional
\begin{eqnarray}
F[p] & ~\doteq~ & H(\mathcal{X}\mid\Theta)-H(\Theta)
-\lambda \left(\int_\Theta p(\boldsymbol{\theta})\,|d\boldsymbol{\theta}|-1\right)
\\& ~=~ &
\lambda+\int_\Theta\left\{
H(\mathcal{X}\mid\boldsymbol{\theta})+\log p(\boldsymbol{\theta})-\lambda
\right\}\,p(\boldsymbol{\theta})\,|d\boldsymbol{\theta}|\,.
\end{eqnarray}
By the calculus of variations, $F$ is minimised when $\delta F=0$, i.e.
\begin{eqnarray}
H(\mathcal{X}\mid\boldsymbol{\theta})+\log p(\boldsymbol{\theta}) 
& ~=~ & \lambda\,.
\end{eqnarray}
Note that if the logarithms are in base $b$, then $\log_b z=\ln z/\ln b$.
Hence, multiplying boths sides by $\ln b$ and exponentiating, we obtain
\begin{eqnarray}
p(\boldsymbol{\theta}) & = & k\,e^{-\ln b\,H(\mathcal{X}\mid\boldsymbol{\theta})}
\end{eqnarray}
as the *minimum information prior*. We leave out proof that this is indeed the minimum.

## Examples of Minimum Information

### Gaussian distribution prior

Following Zellner [[5]](#References "Reference: An introduction to Bayesian inference in econometrics"), we consider the univariate Gaussian distribution
\begin{eqnarray}
p(x\mid\mu,\sigma) & = & \frac{1}{\sqrt{2\pi\sigma^2}}
e^{-\frac{(x-\mu)^2}{2\sigma^2}}\,,
\end{eqnarray}
such that
\begin{eqnarray}
\ln p(x\mid\mu,\sigma) & = & -\frac{1}{2}\ln(2\pi\sigma^2)
-\frac{(x-\mu)^2}{2\sigma^2}\,,
\\
\Rightarrow H(X\mid\mu,\sigma) & = &
\int_{-\infty}^{\infty}\left[
\frac{1}{2}\ln(2\pi\sigma^2)
+\frac{(x-\mu)^2}{2\sigma^2}
\right]\,p(x\mid\mu,\sigma)\,dx
\\& = &
\frac{1}{2}\ln(2\pi\sigma^2)+\frac{1}{2\sigma^2}\mathbb{V}(X)
\\& = &
\frac{1}{2}\left[1+\ln(2\pi\sigma^2)\right]
~=~\frac{1+\ln(2\pi)}{2}+\ln\sigma
\,,
\end{eqnarray}
since $\mathbb{V}[X]=\sigma^2$.
The
[minimum information](#Minimum-information-principle "Section: Minimum information principle") prior is therefore (with logarithmic base $b=e$) given by
\begin{eqnarray}
p(\mu,\sigma) & = & k\,e^{-H(X\mid\boldsymbol{\theta})}~=~\frac{k'}{\sigma}\,.
\end{eqnarray}
This agrees with the joint prior found using
[transformation group invariance](#Nonlinear-rescaling-plus-relocation "Section: Nonlinear rescaling plus relocation").

However, note that if we reparameterised the joint distribution in terms of the
variance $\sigma^2$, then we would instead obtain
\begin{eqnarray}
p(\mu,\sigma^2) & = & k\,e^{-H(X\mid\boldsymbol{\theta})}
~=~\frac{k'}{\sqrt{\sigma^2}}\,.
\end{eqnarray}
This now disagrees with the joint prior found using
[transformation group invariance](#Nonlinear-rescaling-plus-relocation "Section: Nonlinear rescaling plus relocation"), which would be
\begin{eqnarray}
p(\mu,\sigma^2) & = & \frac{k'}{\sigma^2}\,,
\end{eqnarray}
since in that section we deduced that $p(\mu)\propto 1$ and $p(\sigma^n)\propto(\sigma^n)^{-1}$.
We therefore conclude that the 
[minimum information](#Minimum-information-principle "Section: Minimum information principle") prior is **not** invariant to transformation.

### Exponential distribution prior

We turn now to the exponential distribution
\begin{eqnarray}
& p(t\mid\lambda) & ~=~ \lambda e^{-\lambda t}\,,
\\\Rightarrow~ &
\ln p(t\mid\lambda) & ~=~ \ln\lambda-\lambda t\,,
\\\Rightarrow~ &
H(T\mid\lambda) & ~=~ \int_{0}^{\infty}\left[
\lambda t-\ln\lambda
\right]\,p(t\mid\lambda)\,dt
\\&& ~=~
\lambda E[T\mid\lambda]-\ln\lambda~=~1-\ln\lambda\,,
\end{eqnarray}
since we know that $E[T\mid\lambda]=\lambda^{-1}$.
The
[minimum information](#Minimum-information-principle "Section: Minimum information principle") prior is therefore
\begin{eqnarray}
p(\lambda) & = & k\,e^{-H(T\mid\lambda)}~=~k'\lambda\,.
\end{eqnarray}
However, this form of prior is unexpected, since we saw that 
[invariance to scaling](#Time-scale-invariance "Section: Time-scale invariance")
led to the prior $p(\lambda)\propto\lambda^{-1}$.

In fact, if we reparameterise using the time-scale 
$\tau\doteq\lambda^{-1}$, then we instead obtain
\begin{eqnarray}
& p(t\mid\tau) & ~=~ \frac{1}{\tau} e^{-\frac{t}{\tau}}\,,
\\\Rightarrow~ &
\ln p(t\mid\tau) & ~=~ -\ln\tau-\frac{t}{\tau}\,,
\\\Rightarrow~ &
H(T\mid\tau) & ~=~ \int_{0}^{\infty}\left[
\frac{t}{\tau}+\ln\tau
\right]\,p(t\mid\tau)\,dt
\\&& ~=~
\frac{1}{\tau}\,E[T\mid\tau]+\ln\tau~=~1+\ln\tau\,,
\end{eqnarray}
since $E[T\mid\tau]=\tau$.
The minimum information prior is now
\begin{eqnarray}
p(\tau) & = & k\,e^{-H(T\mid\tau)}~=~\frac{k'}{\tau}\,.
\end{eqnarray}
This does now has the form of prior expected from 
[invariance to scaling](#Time-scale-invariance "Section: Time-scale invariance").
Hence, since the two priors under transformation have completely different forms, we once again conclude that the 
[minimum information](#Minimum-information-principle "Section: Minimum information principle") prior is not invariant to transformation.
